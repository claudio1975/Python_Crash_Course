## 1. Strings

### What is a String?

A **string** is a sequence of characters, basically any text you want to work with. Think of it as a **necklace of beads**, where each bead is a single character (letter, number, space, or symbol).

```
H  e  l  l  o
0  1  2  3  4   ← index (position of each character)
```

📌 **Key fact:** Strings are **immutable**, once created, you can't change individual characters. You create a new string instead.


````python
# ── Creating Strings ─────────────────────────────────────
single_quotes = 'Hello, World!'
double_quotes = "Hello, World!"     # Same result, either style is fine
multiline     = """This string
spans multiple
lines!"""

print(single_quotes)
print(double_quotes)
print(multiline)
````

````python
# ── Accessing Characters (Indexing) ──────────────────────
# Think of indexes as the seat numbers in a cinema row
name = "Python"
#        P  y  t  h  o  n
#        0  1  2  3  4  5   ← positive index
#       -6 -5 -4 -3 -2 -1   ← negative index (from the end)

print("First character:",  name[0])   
print("Third character:",  name[2])   
print("Last character:",   name[-1])  
print("Second from end:",  name[-2])  
````

````python
# ── Slicing — extracting a portion of the string ─────────
# string[start:stop:step]
# 'stop' is EXCLUDED (like a bus stop you didn't get on)
word = "Programming"

print(word[0:7])    # 'Program'  → chars from index 0 to 6
print(word[3:])     # 'gramming' → from index 3 to the end
print(word[:4])     # 'Prog'     → from start up to index 3
print(word[::-1])   # reverse the whole string!
````

````python
# ── Common String Methods ─────────────────────────────────
sentence = "  hello, python world!  "

print(sentence.strip())          # Remove spaces at edges
print(sentence.strip().upper())  # ALL CAPS
print(sentence.strip().title())  # Title Case

message = "I love cats and cats love me"
print(message.replace("cats", "dogs"))   # Replace text
print(message.count("cats"))             # Count occurrences
print(message.split())                   # Split into a list of words

# Check content
print("123".isdigit())   # True — only digits
print("abc".isalpha())   # True — only letters
print("python" in message)  # False — check if substring exists
````

````python
# ── String Formatting (f-strings — the modern way!) ──────
# f-strings let you embed variables directly inside a string
name = "Alice"
age  = 30
city = "Milan"

# Old way:
print("My name is " + name + " and I am " + str(age) + " years old.")

# f-string way (much cleaner!):
print(f"My name is {name} and I am {age} years old.")
print(f"I live in {city.upper()}.")
print(f"Next year I'll be {age + 1}.")   # you can do math inside {}
````

````python
# ── Joining and Concatenating ─────────────────────────────
first = "Data"
second = "Science"

# Concatenation with +
combined = first + " " + second
print(combined)

# Repetition with *
print("Ha" * 3)

# join() — glue a list of strings with a separator
words = ["Python", "is", "awesome"]
sentence = " ".join(words)
print(sentence)

csv_line = ",".join(["Alice", "30", "Milan"])
print(csv_line)
````

**🧠 String Cheat Sheet:**

| Method | What it does | Example |
|---|---|---|
| `.upper()` | Converts to uppercase | `"hello".upper()` → `"HELLO"` |
| `.lower()` | Converts to lowercase | `"HELLO".lower()` → `"hello"` |
| `.strip()` | Removes leading/trailing spaces | `"  hi  ".strip()` → `"hi"` |
| `.replace(a, b)` | Replaces `a` with `b` | `"cat".replace("c","b")` → `"bat"` |
| `.split()` | Splits into a list of words | `"a b".split()` → `["a","b"]` |
| `.join(list)` | Joins list into a string | `"-".join(["a","b"])` → `"a-b"` |
| `len()` | Length (number of characters) | `len("hello")` → `5` |
| `in` | Check if substring exists | `"py" in "python"` → `True` |

## 📋 2. Lists

### What is a List?

A **list** is an **ordered, changeable collection** that can hold any mix of items. Imagine it as a **shopping list**: you can add items, remove them, change the order, and it remembers the sequence.

```python
my_list = [item1, item2, item3, ...]
```

✅ **Lists can hold anything:** numbers, strings, booleans, even other lists!  
✅ **Lists are ordered:** items always stay in the order you put them  
✅ **Lists are mutable:** you can change, add, and remove items  



````python
# ── Creating Lists ────────────────────────────────────────
fruits     = ["apple", "banana", "cherry"]
numbers    = [10, 20, 30, 40, 50]
mixed      = ["Alice", 25, True, 3.14]   # Any mix of types!
empty_list = []

print(fruits)
print(f"Number of fruits: {len(fruits)}")
print(f"Type: {type(fruits)}")
````

````python
# ── Accessing Items (same indexing as strings!) ───────────
colors = ["red", "green", "blue", "yellow", "purple"]
#          0       1        2       3         4
#         -5      -4       -3      -2        -1

print(colors[0])    # red    (first item)
print(colors[-1])   # purple (last item)
print(colors[1:4])  # slice: ['green', 'blue', 'yellow']
````

````python
# ── Modifying Lists ───────────────────────────────────────
shopping = ["milk", "bread", "eggs"]

# Change an item
shopping[1] = "whole wheat bread"   # Update item at index 1
print("After update:", shopping)

# Add items
shopping.append("butter")           # Add to the END
shopping.insert(1, "orange juice")  # Insert at a specific position
print("After adding:", shopping)

# Remove items
shopping.remove("eggs")             # Remove by VALUE
popped = shopping.pop()             # Remove and return the LAST item
print("After removing:", shopping)
print("Popped item:", popped)
````

````python
# ── Useful List Operations ────────────────────────────────
scores = [85, 42, 97, 61, 73, 55, 90]

print("Original:", scores)
print("Sorted (ascending):",  sorted(scores))
print("Sorted (descending):", sorted(scores, reverse=True))
print("Min:", min(scores))
print("Max:", max(scores))
print("Sum:", sum(scores))
print("Average:", sum(scores) / len(scores))

# Check if item is in the list
print("Is 97 in scores?", 97 in scores)
print("Is 100 in scores?", 100 in scores)
````

````python
# ── Iterating Through a List ──────────────────────────────
students = ["Alice", "Bob", "Charlie", "Diana"]

# Simple loop
for student in students:
    print(f"Hello, {student}!")

print()

# Loop with index using enumerate()
for i, student in enumerate(students, start=1):
    print(f"{i}. {student}")
````

````python
# ── Combining Lists ───────────────────────────────────────
list_a = [1, 2, 3]
list_b = [4, 5, 6]

# Concatenation
combined = list_a + list_b
print("Combined:", combined)

# Extend (adds all items from another list)
list_a.extend(list_b)
print("Extended:", list_a)

# Nested lists (a list of lists — like a table!)
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Row 1:", matrix[0])
print("Item at row 2, col 3:", matrix[1][2])  # 6
````

**🧠 List Methods Cheat Sheet:**

| Method | What it does |
|---|---|
| `.append(x)` | Add `x` to the end |
| `.insert(i, x)` | Insert `x` at position `i` |
| `.remove(x)` | Remove first occurrence of `x` |
| `.pop()` | Remove and return last item |
| `.sort()` | Sort the list in place |
| `.reverse()` | Reverse the list in place |
| `.count(x)` | Count how many times `x` appears |
| `.index(x)` | Find the index of `x` |
| `.clear()` | Remove all items |

## 🔒 3. Tuples

### What is a Tuple?

A **tuple** is like a list, but it's **locked**, you can't change it once created. Think of a tuple as a **published book**: the content is fixed, you can read it, but you can't edit the pages.

```python
my_tuple = (item1, item2, item3)
```

🔒 **Immutable:** items cannot be added, removed, or changed  
⚡ **Faster than lists** for reading data  
✅ **Good for data that should not change:** coordinates, RGB colors, database records  


````python
# ── Creating Tuples ───────────────────────────────────────
coordinates   = (48.8566, 2.3522)        # Latitude, Longitude of Paris
rgb_red       = (255, 0, 0)             # Red color in RGB
person        = ("Alice", 30, "Milan")  # Name, age, city
single_item   = (42,)                   # Note the comma! Without it, it's just a number

print(coordinates)
print(rgb_red)
print(f"Type: {type(coordinates)}")
````

````python
# ── Accessing Tuple Items (same as lists) ─────────────────
person = ("Alice", 30, "Software Engineer", "Milan")

print("Name:",       person[0])
print("Age:",        person[1])
print("Last item:",  person[-1])
print("Slice:",      person[1:3])  # (30, 'Software Engineer')

# You CANNOT change a tuple:
try:
    person[0] = "Bob"  # This will cause an error!
except TypeError as e:
    print(f"\nError: {e}")
    print("Tuples are read-only!")
````

````python
# ── Tuple Unpacking — a very Pythonic feature! ────────────
# You can assign each item to a separate variable in one line
coordinates = (48.8566, 2.3522)
latitude, longitude = coordinates
print(f"Latitude: {latitude}, Longitude: {longitude}")

# Works for any tuple
name, age, city = ("Bob", 25, "Rome")
print(f"{name} is {age} years old and lives in {city}")

# Swap two variables elegantly!
a, b = 10, 20
print(f"Before swap: a={a}, b={b}")
a, b = b, a    # Swap!
print(f"After swap:  a={a}, b={b}")
````

````python
# ── Tuples in Functions — returning multiple values ───────
def get_min_max(numbers):
    """Returns two values as a tuple"""
    return min(numbers), max(numbers)  # Python packs these as a tuple

data = [5, 3, 9, 1, 7, 2, 8]
minimum, maximum = get_min_max(data)  # Unpack the returned tuple
print(f"Min: {minimum}, Max: {maximum}")

# Tuple methods
fruits = ("apple", "banana", "apple", "cherry", "apple")
print(f"Count of 'apple': {fruits.count('apple')}")
print(f"Index of 'banana': {fruits.index('banana')}")
````

**🆚 List vs Tuple — when to use which?**

| Feature | List `[]` | Tuple `()` |
|---|---|---|
| **Mutable?** | ✅ Yes — can change | ❌ No — fixed forever |
| **Speed** | Slightly slower | ⚡ Faster |
| **Use for** | Dynamic collections | Fixed data, coordinates, records |
| **Example** | Shopping cart | GPS coordinates, RGB colors |

> 💡 **Rule of thumb:** if the data *should* change, use a list. If the data *should not* change, use a tuple.

## 📖 4. Dictionaries

### What is a Dictionary?

A **dictionary** stores data as **key-value pairs**, like a real dictionary where you look up a word (key) to find its definition (value). Each key must be unique, just like words in a dictionary.

```python
my_dict = {"key1": value1, "key2": value2}
```

🔑 **Keys** must be unique and immutable (strings, numbers, tuples)  
💾 **Values** can be anything (strings, numbers, lists, other dicts!)  
⚡ **Very fast lookup** — finding a value by key is instant  


````python
# ── Creating Dictionaries ─────────────────────────────────
student = {
    "name": "Alice",
    "age": 21,
    "major": "Computer Science",
    "gpa": 3.8,
    "is_active": True
}

print(student)
print(f"Type: {type(student)}")
print(f"Keys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
````

````python
# ── Accessing Values ──────────────────────────────────────
student = {"name": "Alice", "age": 21, "city": "Milan"}

# Method 1: square brackets (raises error if key missing!)
print(student["name"])    # Alice

# Method 2: .get() — safer, returns None if key is missing
print(student.get("age"))         # 21
print(student.get("phone"))       # None (no error!)
print(student.get("phone", "N/A"))  # Default value if missing

# Check if a key exists
print("name" in student)    # True
print("email" in student)   # False
````

````python
# ── Adding and Modifying Items ────────────────────────────
person = {"name": "Bob", "age": 25}

# Add a new key-value pair
person["city"] = "Turin"
person["email"] = "bob@example.com"
print("After adding:", person)

# Modify an existing value
person["age"] = 26
print("After updating age:", person)

# Update with another dictionary
person.update({"hobby": "cycling", "city": "Rome"})  # Also overwrites existing keys
print("After update():", person)

# Remove a key
removed = person.pop("email")  # Removes and returns the value
print(f"Removed: {removed}")
print("After removing email:", person)
````

````python
# ── Iterating Through a Dictionary ───────────────────────
product = {
    "name":     "Laptop",
    "brand":    "TechBrand",
    "price":    999.99,
    "in_stock": True
}

# Loop through keys
print("Keys:", list(product.keys()))

# Loop through values
print("Values:", list(product.values()))

# Loop through key-value pairs (most useful!)
print("\nProduct details:")
for key, value in product.items():
    print(f"  {key}: {value}")
````

````python
# ── Nested Dictionaries — dict inside a dict ──────────────
# Perfect for representing complex real-world data!
company = {
    "name": "TechCorp",
    "employees": {
        "alice": {"role": "Developer", "salary": 65000},
        "bob":   {"role": "Designer",  "salary": 58000},
    }
}

# Access nested values
print(company["name"])
print(company["employees"]["alice"]["role"])
print(company["employees"]["bob"]["salary"])
````

**🧠 Dictionary Cheat Sheet:**

| Operation | Code |
|---|---|
| Get value | `d["key"]` or `d.get("key")` |
| Add / Update | `d["key"] = value` |
| Remove key | `d.pop("key")` |
| All keys | `d.keys()` |
| All values | `d.values()` |
| All pairs | `d.items()` |
| Check key exists | `"key" in d` |
| Number of items | `len(d)` |

## 🎯 5. Sets

### What is a Set?

A **set** is a collection of **unique items** with **no duplicates**, and items have **no order**. Think of it like a bag of poker chips: each chip is unique, and the order you put them in doesn't matter.

```python
my_set = {item1, item2, item3}
```

🎯 **Unique items only** — duplicates are automatically removed  
🔀 **Unordered** — items have no guaranteed position  
⚡ **Super fast** at checking if something is in the set  
🔢 **Great for math operations** like union, intersection, difference  


````python
# ── Creating Sets ─────────────────────────────────────────
fruits     = {"apple", "banana", "cherry"}
with_dupes = {1, 2, 2, 3, 3, 3, 4}   # Duplicates are silently removed!
from_list  = set(["a", "b", "a", "c", "b"])  # Convert list to set

print(fruits)
print("With duplicates removed:", with_dupes)
print("From list:", from_list)
print(f"Type: {type(fruits)}")

# ⚠️ Empty set: use set(), NOT {} — that creates an empty DICTIONARY!
empty = set()
print(f"Empty set type: {type(empty)}")
````

````python
# ── Adding and Removing from Sets ────────────────────────
colors = {"red", "green", "blue"}

colors.add("yellow")     # Add one item
print("After add:", colors)

colors.discard("green")  # Remove item (no error if it doesn't exist)
print("After discard:", colors)

colors.remove("red")     # Remove item (raises KeyError if missing)
print("After remove:", colors)

# Check membership — this is where sets shine!
print("Is 'blue' in colors?", "blue" in colors)
````

````python
# ── Set Math Operations ───────────────────────────────────
# This is where sets really shine — like a Venn diagram!
python_students  = {"Alice", "Bob", "Charlie", "Diana"}
java_students    = {"Bob", "Diana", "Eve", "Frank"}

# UNION — students who take at least one course
all_students = python_students | java_students   # or: .union()
print("Union (all students):", all_students)

# INTERSECTION — students in BOTH courses
both_courses = python_students & java_students   # or: .intersection()
print("Intersection (both courses):", both_courses)

# DIFFERENCE — students ONLY in Python course
only_python = python_students - java_students    # or: .difference()
print("Difference (only Python):", only_python)

# SYMMETRIC DIFFERENCE — students in one course but NOT both
one_course_only = python_students ^ java_students  # or: .symmetric_difference()
print("Symmetric difference (one course only):", one_course_only)
````

````python
# ── Practical Use: Removing Duplicates from a List ────────
# This is one of the most common reasons to use sets!

visitor_log = ["alice", "bob", "alice", "charlie", "bob", "alice", "diana"]
print("Full log:", visitor_log)
print(f"Total entries: {len(visitor_log)}")

unique_visitors = list(set(visitor_log))  # Convert to set (removes dupes), back to list
print("\nUnique visitors:", sorted(unique_visitors))
print(f"Unique count: {len(unique_visitors)}")
````

**🧠 Set Operations Cheat Sheet:**

| Operation | Symbol | Method |
|---|---|---|
| Union (A or B) | `A \| B` | `A.union(B)` |
| Intersection (A and B) | `A & B` | `A.intersection(B)` |
| Difference (A not in B) | `A - B` | `A.difference(B)` |
| Symmetric difference | `A ^ B` | `A.symmetric_difference(B)` |
| Is subset? | `A <= B` | `A.issubset(B)` |
| Is superset? | `A >= B` | `A.issuperset(B)` |

## 🗂️ 6. Arrays

### What is an Array?

An **array** is like a list, but it can **only hold items of the same type** (all integers, or all floats, etc.). Think of it as a **row of identical storage boxes**, every box holds the same kind of item.

Python's built-in `array` module provides this. For data science, **NumPy arrays** are far more common, but the concept is the same.

🔢 **Typed:** all items must be the same data type  
💾 **Memory efficient:** uses less memory than a list for the same data  
⚡ **Faster math operations** than regular lists  


````python
# ── Using Python's Built-in Array Module ─────────────────
from array import array

# Type codes: 'i' = integer, 'f' = float, 'd' = double
# array(type_code, [initial_values])
int_array   = array('i', [10, 20, 30, 40, 50])
float_array = array('f', [1.5, 2.7, 3.14, 4.0])

print("Integer array:", int_array)
print("Float array:",   float_array)
print(f"Type: {type(int_array)}")
print(f"First item: {int_array[0]}")
print(f"Length: {len(int_array)}")
````

````python
# ── Array Operations ──────────────────────────────────────
from array import array
temps = array('f', [22.5, 18.3, 25.1, 19.8, 23.4])

# Append and insert
temps.append(21.0)
temps.insert(0, 17.5)
print("After modifications:", list(temps))

# Remove
temps.remove(17.5)
print("After removal:", list(temps))

# Convert to list for more operations
temps_list = list(temps)
print(f"Average temp: {sum(temps_list)/len(temps_list):.1f}°C")
print(f"Max: {max(temps_list)}°C")
print(f"Min: {min(temps_list)}°C")
````

````python
# ── Arrays vs Lists: Type Enforcement ────────────────────
from array import array
int_arr = array('i', [1, 2, 3])

# Arrays enforce type!
int_arr.append(4)     # OK — integer
print("After appending 4:", list(int_arr))

try:
    int_arr.append("hello")  # Not OK — string in integer array!
except TypeError as e:
    print(f"Error: {e}")
    print("Arrays enforce a single data type — lists do not!")
````

**🆚 List vs Array — when to use which?**

| Feature | List | Array |
|---|---|---|
| **Data types** | Any mix | Single type only |
| **Memory** | More memory | Less memory |
| **Math performance** | Slower | Faster |
| **Flexibility** | Very flexible | Less flexible |
| **Best for** | General use, mixed data | Numeric data, large datasets |

> 💡 In data science and machine learning, **NumPy arrays** (not the built-in `array`) are the standard. The concepts are the same, but NumPy is much more powerful.

## ✨ 7. List Comprehension

### What is List Comprehension?

**List comprehension** is a concise, elegant way to create a new list by transforming or filtering an existing one. It's like writing a recipe in a single sentence instead of a full paragraph.

**Long way (without comprehension):**
```python
squares = []
for x in range(10):
    squares.append(x ** 2)
```

**Short way (with comprehension):**
```python
squares = [x ** 2 for x in range(10)]
```

**General syntax:**
```python
[expression for item in iterable if condition]
```


````python
# ── Basic List Comprehension ──────────────────────────────
# Generate squares of numbers 0 to 9
squares = [x ** 2 for x in range(10)]
print("Squares:", squares)

# Convert Celsius to Fahrenheit
celsius = [0, 10, 20, 30, 40, 100]
fahrenheit = [(c * 9/5) + 32 for c in celsius]
print("Fahrenheit:", fahrenheit)

# Make all names uppercase
names = ["alice", "bob", "charlie", "diana"]
upper_names = [name.upper() for name in names]
print("Uppercase:", upper_names)
````

````python
# ── Filtering with Conditions ─────────────────────────────
# The 'if condition' part acts like a filter
numbers = list(range(1, 21))  # 1 to 20

# Keep only even numbers
evens = [n for n in numbers if n % 2 == 0]
print("Even numbers:", evens)

# Keep only numbers divisible by 3
div_by_3 = [n for n in numbers if n % 3 == 0]
print("Divisible by 3:", div_by_3)

# Filter words longer than 4 letters
words = ["cat", "python", "dog", "programming", "fox", "data"]
long_words = [w for w in words if len(w) > 4]
print("Long words:", long_words)
````

````python
# ── Combining Transformation AND Filtering ────────────────
scores = [45, 88, 73, 92, 55, 61, 78, 38, 95]

# Get squared values of only the passing scores (>=60)
passing_squared = [s ** 2 for s in scores if s >= 60]
print("Passing scores squared:", passing_squared)

# Classify scores as pass/fail
results = ["Pass" if s >= 60 else "Fail" for s in scores]
print("Results:", results)
````

````python
# ── Nested List Comprehension ─────────────────────────────
# Create a multiplication table as a nested list
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]

print("5×5 Multiplication Table:")
for row in table:
    print(row)

# Flatten a nested list
nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [item for sublist in nested for item in sublist]
print("\nFlattened:", flat)
````

````python
# ── Dictionary and Set Comprehension ─────────────────────
# The same concept works for dicts and sets!

# Dict comprehension: word → length
words = ["apple", "banana", "cherry", "date"]
word_lengths = {word: len(word) for word in words}
print("Word lengths:", word_lengths)

# Dict comprehension with filtering
scores = {"Alice": 88, "Bob": 45, "Charlie": 92, "Diana": 55}
passing = {name: score for name, score in scores.items() if score >= 60}
print("Passing students:", passing)

# Set comprehension
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_squares = {n ** 2 for n in numbers}
print("Unique squares:", unique_squares)
````

## 🏆 Comprehensive Example: Library Management System

Now let's **bring everything together** in a realistic example that uses all 7 data structures!

We'll build a simple **Library Management System** that manages books, members, and borrowing records.

````python
# ============================================================
# 📚 LIBRARY MANAGEMENT SYSTEM
# ============================================================

from array import array

print("=" * 60)
print("           📚 CITY LIBRARY MANAGEMENT SYSTEM")
print("=" * 60)

# ── 1. STRINGS — Book information ─────────────────────────
library_name    = "City Central Library"
library_address = "123 Book Street, Milan"
motto           = "knowledge is power"

# String formatting and methods
print(f"\nWelcome to the {library_name.upper()}")
print(f"Address: {library_address}")
print(f"Motto: '{motto.title()}'")
print(f"Library name has {len(library_name)} characters")
````

````python
# ── 2. LIST — Catalogue of all books ──────────────────────
# Each book stored as: [title, author, year, copies_available]
catalogue = [
    ["The Great Gatsby",           "F. Scott Fitzgerald", 1925, 3],
    ["To Kill a Mockingbird",      "Harper Lee",          1960, 5],
    ["1984",                        "George Orwell",       1949, 4],
    ["Pride and Prejudice",         "Jane Austen",         1813, 2],
    ["The Catcher in the Rye",     "J.D. Salinger",       1951, 1],
    ["Harry Potter and the Sorcerer\'s Stone", "J.K. Rowling", 1997, 6],
    ["The Hitchhiker's Guide",     "Douglas Adams",       1979, 3],
]

print("\n📋 LIBRARY CATALOGUE")
print("-" * 60)
print(f"{'#':<4} {'Title':<40} {'Year':<6} {'Copies'}")
print("-" * 60)

for i, book in enumerate(catalogue, 1):
    title, author, year, copies = book   # Unpack the list
    print(f"{i:<4} {title:<40} {year:<6} {copies}")

print(f"\nTotal books in catalogue: {len(catalogue)}")
````

````python
# ── 3. TUPLE — Immutable member records ───────────────────
# Member data that shouldn't change: (id, name, join_date)
members = [
    (1001, "Alice Rossi",    "2021-03-15"),
    (1002, "Marco Ferrari",  "2022-07-01"),
    (1003, "Sofia Bianchi",  "2020-11-22"),
    (1004, "Luca Romano",    "2023-01-10"),
    (1005, "Elena Conti",    "2022-09-05"),
]

print("\n👤 REGISTERED MEMBERS")
print("-" * 50)

for member_id, name, join_date in members:   # Tuple unpacking!
    print(f"  ID: {member_id} | {name:<20} | Joined: {join_date}")

print(f"\nTotal members: {len(members)}")
````

````python
# ── 4. DICTIONARY — Borrowing records ─────────────────────
# member_id → list of books currently borrowed
borrowing_records = {
    1001: ["1984", "The Great Gatsby"],
    1002: ["Harry Potter and the Sorcerer's Stone"],
    1003: [],   # No books borrowed currently
    1004: ["Pride and Prejudice", "To Kill a Mockingbird"],
    1005: ["The Hitchhiker's Guide"],
}

# Book details for quick lookup
book_info = {
    book[0]: {"author": book[1], "year": book[2], "copies": book[3]}
    for book in catalogue
}

print("\n📖 BORROWING RECORDS")
print("-" * 55)

for member in members:
    member_id, name, _ = member  # Ignore join_date with _
    borrowed = borrowing_records.get(member_id, [])

    if borrowed:
        print(f"  {name} ({member_id}): borrowed {len(borrowed)} book(s)")
        for title in borrowed:
            print(f"    → '{title}'")
    else:
        print(f"  {name} ({member_id}): no books borrowed")
````

````python
# ── 5. SET — Unique books currently on loan ───────────────
# Collect ALL borrowed books (from all members) as a set
all_borrowed = set()
for books in borrowing_records.values():
    all_borrowed.update(books)  # Add all books from each member

# Set of all available book titles
all_titles = {book[0] for book in catalogue}

# Set operations!
available = all_titles - all_borrowed      # Books NOT currently borrowed
on_loan   = all_titles & all_borrowed      # Books that ARE borrowed

print("\n🔍 AVAILABILITY REPORT (using Sets)")
print("-" * 55)
print(f"Total titles: {len(all_titles)}")
print(f"Titles with at least 1 copy on loan: {len(on_loan)}")
print(f"Titles fully available: {len(available)}")

print("\n  On loan:")
for title in sorted(on_loan):
    print(f"    📤 {title}")

print("\n  Fully available:")
for title in sorted(available):
    print(f"    ✅ {title}")
````

````python
# ── 6. ARRAY — Storing publication years numerically ──────
years = array('i', [book[2] for book in catalogue])

years_list = list(years)

print("\n📅 PUBLICATION YEAR STATISTICS (using Array)")
print("-" * 55)
print(f"Years: {years_list}")
print(f"Oldest book year: {min(years_list)}")
print(f"Newest book year: {max(years_list)}")
print(f"Average publication year: {sum(years_list)//len(years_list)}")
````

````python
# ── 7. LIST COMPREHENSION — Smart analytics ───────────────

# Books published after 1950
modern_books = [
    f"{book[0]} ({book[2]})"
    for book in catalogue
    if book[2] >= 1950
]

# Books with multiple copies available
well_stocked = [
    book[0]
    for book in catalogue
    if book[3] >= 4
]

# All authors (no duplicates via set comprehension)
authors = {book[1] for book in catalogue}

# Member activity: member_name → number of books borrowed
activity = {
    name: len(borrowing_records.get(mid, []))
    for mid, name, _ in members
}

print("\n✨ ANALYTICS (using List Comprehension)")
print("-" * 55)

print(f"\n📚 Books published after 1950:")
for book in modern_books:
    print(f"    • {book}")

print(f"\n📦 Well-stocked books (4+ copies):")
for book in well_stocked:
    print(f"    • {book}")

print(f"\n✍️  Unique authors: {len(authors)}")
for author in sorted(authors):
    print(f"    • {author}")

print(f"\n📊 Member activity (books borrowed):")
for name, count in sorted(activity.items(), key=lambda x: x[1], reverse=True):
    bar = "█" * count
    print(f"    {name:<20} {bar} ({count})")
````

````python
# ── FINAL SUMMARY ─────────────────────────────────────────
total_borrowed_books = sum(len(v) for v in borrowing_records.values())
most_active = max(activity.items(), key=lambda x: x[1])

print("\n" + "=" * 60)
print("                   📊 LIBRARY SUMMARY")
print("=" * 60)
print(f"  📚 Total titles in catalogue:   {len(catalogue)}")
print(f"  👤 Registered members:          {len(members)}")
print(f"  📤 Books currently on loan:     {total_borrowed_books}")
print(f"  ✅ Titles fully available:      {len(available)}")
print(f"  🏆 Most active member:          {most_active[0]} ({most_active[1]} books)")
print(f"  📅 Oldest book:                 {min(years_list)}")
print(f"  📅 Newest book:                 {max(years_list)}")
print("=" * 60)
print("\n📖 All data structures used in this example:")
data_structures_used = [
    ("String",            "Library name, address, book titles"),
    ("List",              "Book catalogue, borrowed book lists"),
    ("Tuple",             "Member records (immutable data)"),
    ("Dictionary",        "Borrowing records, book info lookup"),
    ("Set",               "Unique available/borrowed titles"),
    ("Array",             "Publication years (numeric storage)"),
    ("List Comprehension","Analytics, filtering, transformations"),
]
for ds, usage in data_structures_used:
    print(f"  ✅ {ds:<22} → {usage}")
print("\n" + "=" * 60)
````
