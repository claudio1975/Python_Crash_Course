## 4. Variables

### What are Variables?

Think of variables as **labeled boxes** where you can store information. You give the box a name, and Python remembers what's inside.

```
┌──────────────┐
│  age = 25    │  <- This box is labeled "age" and contains 25
└──────────────┘
```

````python
# Creating variables - just use = to assign a value
age = 25
name = "Alice"
height = 5.6
is_student = True

# Using variables
print("Name:", name)
print("Age:", age)
print("Height:", height, "feet")
print("Student:", is_student)
````

### Variable Naming Rules:

✅ **Good variable names:**
```python
student_name
age_in_years
total_score
_private_var
```

❌ **Bad variable names:**
```python
2students      # Can't start with a number
student-name   # Can't use hyphens
class          # Can't use keywords
my name        # Can't have spaces
```

### Rules:
1. Start with a letter (a-z, A-Z) or underscore (_)
2. Can contain letters, numbers, and underscores
3. **No spaces** - use underscore instead: `student_name`
4. **Case-sensitive** - `age` and `Age` are different variables
5. Can't use Python keywords (like `if`, `for`, `while`)

````python
# Variables can change
score = 10
print("Initial score:", score)

score = 20  # Changed the value
print("New score:", score)

score = score + 5  # Use the last value to calculate new value
print("Final score:", score)
````

````python
# Multiple assignments
x, y, z = 1, 2, 3
print("x =", x)
print("y =", y)
print("z =", z)

# Same value to multiple variables
a = b = c = 100
print("a, b, c =", a, b, c)
````

## 5. Operators

Operators are symbols that perform operations on values and variables. Think of them as the **mathematical symbols** you use in calculations.

### 5.1 Arithmetic Operators (Math)

These perform basic math operations:

````python
a = 10
b = 3

print("Addition:", a + b)        # 13
print("Subtraction:", a - b)     # 7
print("Multiplication:", a * b)  # 30
print("Division:", a / b)        # 3.333...
print("Floor Division:", a // b) # 3 (rounds down)
print("Modulus (remainder):", a % b)  # 1
print("Exponentiation:", a ** b) # 1000 (10 to the power of 3)
````

### 5.2 Comparison Operators (Comparing Values)

These compare two values and return `True` or `False`:

````python
x = 10
y = 5

print("Equal:", x == y)              # False
print("Not equal:", x != y)          # True
print("Greater than:", x > y)        # True
print("Less than:", x < y)           # False
print("Greater or equal:", x >= 10)  # True
print("Less or equal:", y <= 5)      # True
````

### 5.3 Logical Operators (Combining Conditions)

These combine multiple conditions:

````python
age = 20
has_license = True

# AND - both must be True
can_drive = age >= 18 and has_license
print("Can drive:", can_drive)  # True

# OR - at least one must be True
is_minor = age < 18 or age > 65
print("Is minor or senior:", is_minor)  # False

# NOT - reverses True/False
is_adult = not (age < 18)
print("Is adult:", is_adult)  # True
````

### 5.4 Assignment Operators (Shortcuts)

These are shorthand ways to update variables:

````python
count = 10

count += 5   # Same as: count = count + 5
print("After += 5:", count)  # 15

count -= 3   # Same as: count = count - 3
print("After -= 3:", count)  # 12

count *= 2   # Same as: count = count * 2
print("After *= 2:", count)  # 24

count /= 4   # Same as: count = count / 4
print("After /= 4:", count)  # 6.0
````

## 7. Data Types

Data types tell Python what **kind of information** you're working with. Python automatically figures out the type based on the value you assign.

### Main Data Types:

### 7.1 Numbers

````python
# Integer (int) - whole numbers
age = 25
temperature = -10
count = 0

print("Age:", age, "- Type:", type(age))

# Float - decimal numbers
height = 5.9
pi = 3.14159
price = 19.99

print("Height:", height, "- Type:", type(height))
````

### 7.2 Strings (Text)

````python
# String (str) - text
name = "Alice"
message = 'Hello, World!'  # Single or double quotes work
multiline = """This is
a multi-line
string"""

print("Name:", name, "- Type:", type(name))

# String operations
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name  # Concatenation
print("Full name:", full_name)

# String repetition
print("Ha" * 3)  # HaHaHa

# String length
print("Length of name:", len(name))
````

### 7.3 Boolean (True/False)

````python
# Boolean (bool) - True or False
is_student = True
is_working = False
has_license = True

print("Is student:", is_student, "- Type:", type(is_student))

# Booleans from comparisons
age = 20
is_adult = age >= 18
print("Is adult:", is_adult)
````

### 7.4 Lists (Ordered Collections)

````python
# List - ordered collection of items
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]
mixed = ["Alice", 25, True, 5.6]  # Can mix types!

print("Fruits:", fruits, "- Type:", type(fruits))

# Accessing items (indexing starts at 0)
print("First fruit:", fruits[0])   # apple
print("Second fruit:", fruits[1])  # banana
print("Last fruit:", fruits[-1])   # orange

# Modifying lists
fruits.append("grape")  # Add item
print("After adding:", fruits)

fruits[0] = "mango"  # Change item
print("After changing:", fruits)
````

### 7.5 Tuples (Immutable Collections)

````python
# Tuple - like a list but cannot be changed
coordinates = (10, 20)
rgb_color = (255, 0, 128)

print("Coordinates:", coordinates, "- Type:", type(coordinates))

# Accessing items
print("X:", coordinates[0])
print("Y:", coordinates[1])

# Tuples cannot be modified
# coordinates[0] = 15  # This would cause an error!
````

### 7.6 Dictionaries (Key-Value Pairs)

````python
# Dictionary - stores key-value pairs
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A",
    "is_active": True
}

print("Student:", student, "- Type:", type(student))

# Accessing values by key
print("Name:", student["name"])
print("Age:", student["age"])

# Adding or modifying
student["major"] = "Computer Science"
student["age"] = 21
print("Updated student:", student)
````

### 7.7 Sets (Unique Collections)

````python
# Set - unordered collection of unique items
colors = {"red", "green", "blue"}
numbers = {1, 2, 3, 2, 1}  # Duplicates are removed!

print("Colors:", colors, "- Type:", type(colors))
print("Numbers (duplicates removed):", numbers)

# Set operations
colors.add("yellow")
print("After adding yellow:", colors)
````

### Type Conversion

You can convert between types:

````python
# String to number
age_text = "25"
age_number = int(age_text)
print("Converted to int:", age_number, type(age_number))

# Number to string
score = 95
score_text = str(score)
print("Converted to string:", score_text, type(score_text))

# Float to int (removes decimal)
price = 19.99
rounded = int(price)
print("Converted to int:", rounded)

# Int to float
whole = 10
decimal = float(whole)
print("Converted to float:", decimal)
````

## 8. Conditional Statements

Conditional statements let your program **make decisions**. Think of them as asking questions and doing different things based on the answer.

### The if Statement

Basic structure:
```python
if condition:
    # do this if condition is True
```

**Note the indentation!** Python uses indentation (4 spaces or a tab) to know which code belongs inside the if statement.

````python
# Simple if statement
age = 20

if age >= 18:
    print("You are an adult")
    print("You can vote")

print("This always runs")  # Not indented, so always executes
````

### if-else Statement

Do one thing if True, another if False:

````python
temperature = 25

if temperature > 30:
    print("It's hot outside!")
else:
    print("The weather is nice")

print("Have a good day!")  # Always runs
````

### if-elif-else Statement

Check multiple conditions in order:

````python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your score is {score}")
print(f"Your grade is {grade}")
````

### Nested if Statements

You can put if statements inside other if statements:

````python
age = 20
has_license = True

if age >= 18:
    print("You are old enough to drive")
    
    if has_license:
        print("You can drive!")
    else:
        print("But you need to get a license first")
else:
    print("You are too young to drive")
````

### Using Logical Operators in Conditions

Combine multiple conditions:

````python
age = 20
has_license = True
has_car = False

# Using AND
if age >= 18 and has_license:
    print("You meet the basic requirements to drive")

# Using OR
if has_car or has_license:
    print("You have at least one driving-related item")

# Using NOT
if not has_car:
    print("You don't have a car yet")
````

## 9. Loops

Loops let you **repeat code** multiple times without writing it over and over. Python has two main types of loops.

### 9.1 For Loops

Use a `for` loop when you know **how many times** you want to repeat something, or when you want to go through items in a collection.

#### Looping with range()

````python
# Print numbers 0 to 4
print("Counting from 0 to 4:")
for i in range(5):
    print(i)

print("\nCounting from 1 to 5:")
for i in range(1, 6):
    print(i)

print("\nCounting by 2s:")
for i in range(0, 10, 2):
    print(i)
````

**Understanding range():**
- `range(5)` - numbers from 0 to 4 (5 numbers total)
- `range(1, 6)` - numbers from 1 to 5
- `range(0, 10, 2)` - start at 0, stop before 10, step by 2

#### Looping through Lists

````python
# Loop through each item
fruits = ["apple", "banana", "orange", "grape"]

print("My favorite fruits:")
for fruit in fruits:
    print(f"I like {fruit}")

# Loop with index
print("\nFruits with numbers:")
for i in range(len(fruits)):
    print(f"{i + 1}. {fruits[i]}")

# Better way: enumerate()
print("\nUsing enumerate:")
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}. {fruit}")
````

#### Looping through Strings

````python
# Each character in a string
name = "Python"

print("Letters in Python:")
for letter in name:
    print(letter)
````

### 9.2 While Loops

Use a `while` loop when you want to repeat **until a condition becomes False**. You might not know how many times it will repeat.

````python
# Count from 1 to 5
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # IMPORTANT: Update the counter!

print("Done counting!")
````

### 9.3 Loop Control: break and continue

- **break**: Exit the loop immediately
- **continue**: Skip the rest of this iteration and go to the next one

````python
# Using break
print("Finding first number divisible by 7:")
for num in range(1, 100):
    if num % 7 == 0:
        print(f"Found it: {num}")
        break  # Stop searching

# Using continue
print("\nPrinting only even numbers:")
for num in range(1, 11):
    if num % 2 != 0:  # If odd
        continue  # Skip to next number
    print(num)
````

### 9.4 Nested Loops

Loops inside loops - useful for working with tables, grids, or combinations:

````python
# Multiplication table
print("Multiplication Table (1-5):")
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i} × {j} = {i * j}")
    print()  # Blank line between rows
````

````python
# Pattern printing
print("Star pattern:")
for row in range(1, 6):
    for col in range(row):
        print("*", end="")
    print()  # New line after each row
````

## 10. Comprehensive Example: Employee Performance & Payroll System

Now let's put everything together in a complete program that uses all the concepts we've learned!

This system manages employee data, calculates salaries with bonuses, and generates performance reports.

````python
# Employee Performance & Payroll System
# This program manages employee information and calculates compensation

print("=" * 60)
print("EMPLOYEE PERFORMANCE & PAYROLL SYSTEM")
print("=" * 60)

# Lists to store employee data
employees = ["Alice Johnson", "Bob Smith", "Charlie Brown", "Diana Prince", "Eve Davis"]
departments = ["Sales", "IT", "Marketing", "Sales", "IT"]
performance_scores = [92, 78, 85, 95, 88]  # Out of 100
base_salaries = [50000, 65000, 55000, 48000, 70000]  # Annual base salary
years_employed = [3, 5, 2, 4, 6]

# Company policies
bonus_rate_excellent = 0.15  # 15% bonus
bonus_rate_good = 0.10       # 10% bonus
bonus_rate_average = 0.05    # 5% bonus

# Display all employees with their information
print("\nEmployee Directory:")
print("-" * 60)

for i in range(len(employees)):
    name = employees[i]
    department = departments[i]
    score = performance_scores[i]
    base_salary = base_salaries[i]
    years = years_employed[i]
    
    # Determine performance rating and bonus
    if score >= 90:
        rating = "Excellent"
        bonus_rate = bonus_rate_excellent
    elif score >= 80:
        rating = "Good"
        bonus_rate = bonus_rate_good
    elif score >= 70:
        rating = "Satisfactory"
        bonus_rate = bonus_rate_average
    else:
        rating = "Needs Improvement"
        bonus_rate = 0.0
    
    # Calculate bonus and total compensation
    bonus = base_salary * bonus_rate
    total_compensation = base_salary + bonus
    
    # Display employee information
    print(f"{i + 1}. {name}")
    print(f"   Department: {department}")
    print(f"   Years Employed: {years} years")
    print(f"   Performance Score: {score}/100 ({rating})")
    print(f"   Base Salary: ${base_salary:,}")
    print(f"   Performance Bonus: ${bonus:,.2f} ({int(bonus_rate*100)}%)")
    print(f"   Total Compensation: ${total_compensation:,.2f}")
    print()

# Calculate company-wide statistics
print("-" * 60)
print("COMPANY STATISTICS")
print("-" * 60)

# Total employees
total_employees = len(employees)
print(f"Total Employees: {total_employees}")

# Average performance score
total_score = 0
for score in performance_scores:
    total_score += score
average_performance = total_score / total_employees
print(f"Average Performance Score: {average_performance:.2f}")

# Total payroll calculation
total_base_payroll = 0
total_bonuses = 0

for i in range(len(employees)):
    base_salary = base_salaries[i]
    score = performance_scores[i]
    
    # Determine bonus rate
    if score >= 90:
        bonus_rate = bonus_rate_excellent
    elif score >= 80:
        bonus_rate = bonus_rate_good
    elif score >= 70:
        bonus_rate = bonus_rate_average
    else:
        bonus_rate = 0.0
    
    total_base_payroll += base_salary
    total_bonuses += base_salary * bonus_rate

total_payroll = total_base_payroll + total_bonuses

print(f"\nPayroll Summary:")
print(f"  Total Base Salaries: ${total_base_payroll:,}")
print(f"  Total Bonuses: ${total_bonuses:,.2f}")
print(f"  Total Annual Payroll: ${total_payroll:,.2f}")

# Performance distribution
rating_counts = {"Excellent": 0, "Good": 0, "Satisfactory": 0, "Needs Improvement": 0}

for score in performance_scores:
    if score >= 90:
        rating_counts["Excellent"] += 1
    elif score >= 80:
        rating_counts["Good"] += 1
    elif score >= 70:
        rating_counts["Satisfactory"] += 1
    else:
        rating_counts["Needs Improvement"] += 1

print("\nPerformance Distribution:")
for rating, count in rating_counts.items():
    if count > 0:
        percentage = (count / total_employees) * 100
        print(f"  {rating}: {count} employees ({percentage:.1f}%)")

# Department analysis
print("\nDepartment Breakdown:")
unique_departments = []
for dept in departments:
    if dept not in unique_departments:
        unique_departments.append(dept)

for dept in unique_departments:
    count = 0
    dept_total_salary = 0
    
    for i in range(len(departments)):
        if departments[i] == dept:
            count += 1
            dept_total_salary += base_salaries[i]
    
    avg_salary = dept_total_salary / count if count > 0 else 0
    print(f"  {dept}: {count} employees, Avg Salary: ${avg_salary:,.2f}")

# Top performers (score >= 90)
print("\nTop Performers (Eligible for Promotion):")
found_top_performers = False

for i in range(len(employees)):
    if performance_scores[i] >= 90:
        print(f"  ⭐ {employees[i]} ({departments[i]}) - {performance_scores[i]} points")
        found_top_performers = True

if not found_top_performers:
    print("  No employees scored 90 or above")

# Employees needing development (score < 70)
print("\nEmployees Requiring Training & Development:")
found_need_training = False

for i in range(len(employees)):
    if performance_scores[i] < 70:
        print(f"  ⚠️  {employees[i]} ({departments[i]}) - {performance_scores[i]} points")
        found_need_training = True

if not found_need_training:
    print("  All employees are performing at satisfactory levels or above!")

# Long-term employees (5+ years)
print("\nLong-Term Employees (5+ years):")
veteran_count = 0

for i in range(len(employees)):
    if years_employed[i] >= 5:
        print(f"  🏆 {employees[i]} - {years_employed[i]} years")
        veteran_count += 1

if veteran_count == 0:
    print("  No employees with 5+ years of service")

print("\n" + "=" * 60)
print("Report Generated Successfully!")
print("=" * 60)
````
