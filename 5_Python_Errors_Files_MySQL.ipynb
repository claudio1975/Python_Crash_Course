## 1. Exception Handling

### What is an Exception?

Imagine you're baking a cake 🎂 and halfway through you realise there's no flour. You have two choices:
- ❌ Crash: throw everything on the floor and give up
- ✅ Handle it: calmly note the problem, go buy flour, and continue

In Python, **exceptions** are unexpected events (errors) that happen while the program is running. **Exception handling** lets your program deal with them gracefully, instead of crashing, it catches the problem and responds intelligently.

### The Four Keywords

| Keyword | Purpose |
|---------|--------|
| `try` | Wrap the risky code here |
| `except` | What to do if an error occurs |
| `else` | Runs only if NO error occurred |
| `finally` | Always runs, perfect for cleanup |

```python
try:
    # risky code
except SomeError:
    # handle the error
else:
    # no error? do this
finally:
    # always runs
```

````python
# The simplest example: dividing by zero
# Without exception handling, Python would crash with an error message

# ❌ This would crash:
# result = 10 / 0

# ✅ With exception handling:
try:
    result = 10 / 0
except ZeroDivisionError:
    print("⚠️  You can't divide by zero! Please use a different number.")

print("The program continues running normally! 🎉")
````

````python
# Using try / except / else / finally together
# Think of it like a full story: attempt → problem? → success? → cleanup

def safe_divide(a, b):
    try:
        result = a / b                 # Risky: might divide by zero
    except ZeroDivisionError:
        print("❌ Cannot divide by zero!")
    except TypeError:
        print("❌ Both values must be numbers!")
    else:
        print(f"✅ {a} / {b} = {result:.4f}")  # Only runs if no error
    finally:
        print("   🏁 Calculation attempt finished.\n")  # Always runs

safe_divide(10, 2)       # Normal case
safe_divide(10, 0)       # Division by zero
safe_divide(10, "hello") # Wrong type
````

````python
# Catching multiple exceptions and getting the error message
# The 'as e' part captures the error message so you can read it

data = ["42", "hello", "7", None, "3.14"]

print("Converting list items to integers:\n")
for item in data:
    try:
        number = int(item)           # int("hello") → ValueError
                                     # int(None)    → TypeError
        print(f"  ✅ '{item}' → {number}")
    except (ValueError, TypeError) as e:
        print(f"  ❌ '{item}' failed: {e}")
````

````python
# The 'raise' keyword intentionally trigger an exception
# Use this when you detect invalid input yourself

def set_age(age):
    if not isinstance(age, int):
        raise TypeError("Age must be a whole number.")
    if age < 0 or age > 150:
        raise ValueError(f"Age {age} is not a realistic value.")
    print(f"✅ Age set to {age}")

# Test it:
for test_value in [25, -5, 200, 3.7]:
    try:
        set_age(test_value)
    except (ValueError, TypeError) as e:
        print(f"❌ Error for {test_value!r}: {e}")
````

## 2. Built-in Exceptions

### Python's Built-in Exception Catalogue

Python comes with many ready-made exception types, each describing a specific kind of problem. Think of them as **category labels** on error reports, the more specific the label, the easier it is to fix the problem.

### The Exception Hierarchy
```
BaseException
 └── Exception
      ├── ArithmeticError
      │    └── ZeroDivisionError
      ├── LookupError
      │    ├── IndexError
      │    └── KeyError
      ├── TypeError
      ├── ValueError
      ├── FileNotFoundError
      ├── AttributeError
      └── ... many more
```

### The Most Common Built-in Exceptions

| Exception | When it happens |
|---|---|
| `ZeroDivisionError` | Dividing by zero |
| `TypeError` | Wrong data type used |
| `ValueError` | Right type, but invalid value |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key doesn't exist |
| `AttributeError` | Object doesn't have that attribute |
| `FileNotFoundError` | File doesn't exist |
| `NameError` | Variable not defined |
| `StopIteration` | Iterator has no more items |

````python
# Demonstrating common built-in exceptions
# Each block shows a different type of error and how to handle it

# 1. ZeroDivisionError
try:
    x = 100 / 0
except ZeroDivisionError:
    print("1. ZeroDivisionError: Can't divide by zero.")

# 2. TypeError
try:
    result = "hello" + 5
except TypeError as e:
    print(f"2. TypeError: {e}")

# 3. ValueError
try:
    num = int("not_a_number")
except ValueError as e:
    print(f"3. ValueError: {e}")

# 4. IndexError
try:
    my_list = [1, 2, 3]
    item = my_list[10]
except IndexError as e:
    print(f"4. IndexError: {e}")

# 5. KeyError
try:
    my_dict = {"name": "Alice", "age": 30}
    city = my_dict["city"]
except KeyError as e:
    print(f"5. KeyError: {e} key not found")

# 6. AttributeError
try:
    number = 42
    number.upper()  # integers don't have .upper()
except AttributeError as e:
    print(f"6. AttributeError: {e}")

# 7. NameError
try:
    print(undefined_variable)
except NameError as e:
    print(f"7. NameError: {e}")

# 8. FileNotFoundError
try:
    f = open("ghost_file.txt", "r")
except FileNotFoundError as e:
    print(f"8. FileNotFoundError: {e}")
````

````python
# You can catch ANY exception with 'Exception' (the parent class)
# Useful for logging unexpected problems, but try to be specific!

risky_operations = [
    lambda: 1 / 0,
    lambda: int("bad"),
    lambda: [1, 2][99],
    lambda: {"a": 1}["z"],
]

for i, op in enumerate(risky_operations, 1):
    try:
        op()
    except Exception as e:
        # type(e).__name__ gives us the exception class name
        print(f"Operation {i}: [{type(e).__name__}] {e}")
````

## 3. User-Defined Exceptions

### Creating Your Own Exceptions

Built-in exceptions are general purpose. But what if you're building a banking app and need an `InsufficientFundsError`? Or a student system that needs a `GradeOutOfRangeError`?

You can create **custom exceptions** by defining a class that inherits from `Exception`. Think of it like creating a new category on your error filing system one that makes sense for your specific application.

**Syntax:**
```python
class MyCustomError(Exception):
    pass
```

### Why Use Custom Exceptions?
- 🏷️ **Clarity**: `InsufficientFundsError` is much more descriptive than a plain `ValueError`
- 🎯 **Precision**: callers can catch your specific error without catching everything else
- 📖 **Documentation**: the exception name itself explains what went wrong

````python
# Simple custom exception inherits from Exception

class InvalidUsernameError(Exception):
    """Raised when a username doesn't meet the requirements."""
    pass

class UsernameTooShortError(InvalidUsernameError):
    """Username is shorter than the minimum length."""
    pass

class UsernameHasSpacesError(InvalidUsernameError):
    """Username contains spaces."""
    pass


def validate_username(username):
    if len(username) < 4:
        raise UsernameTooShortError(
            f"Username '{username}' is too short. Minimum 4 characters."
        )
    if " " in username:
        raise UsernameHasSpacesError(
            f"Username '{username}' cannot contain spaces."
        )
    print(f"✅ Username '{username}' is valid!")


# Test with various usernames
test_usernames = ["alice", "al", "john doe", "bob_123"]

for uname in test_usernames:
    try:
        validate_username(uname)
    except UsernameTooShortError as e:
        print(f"❌ Too short: {e}")
    except UsernameHasSpacesError as e:
        print(f"❌ Has spaces: {e}")
    except InvalidUsernameError as e:
        print(f"❌ Invalid: {e}")
````

````python
# Advanced: Custom exception with extra attributes
# You can attach extra data to your exceptions!

class InsufficientFundsError(Exception):
    """Raised when a bank account doesn't have enough money."""
    
    def __init__(self, balance, amount_needed):
        self.balance = balance
        self.amount_needed = amount_needed
        self.shortfall = amount_needed - balance
        # Call the parent Exception's __init__ with a descriptive message
        super().__init__(
            f"Need €{amount_needed:.2f} but only €{balance:.2f} available. "
            f"Shortfall: €{self.shortfall:.2f}"
        )


class BankAccount:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        print(f"✅ Withdrew €{amount:.2f}. New balance: €{self.balance:.2f}")


account = BankAccount("Alice", balance=100.00)

# Try to withdraw amounts
for amount in [50.00, 30.00, 40.00]:
    try:
        print(f"\nAttempting to withdraw €{amount:.2f}...")
        account.withdraw(amount)
    except InsufficientFundsError as e:
        print(f"❌ Transaction failed: {e}")
        print(f"   Current balance: €{e.balance:.2f}")
        print(f"   You need €{e.shortfall:.2f} more.")
````

## 4. File Handling

### Why Work with Files?

When your program ends, all variables and data disappear from memory. Files are how you **save data permanently** like writing notes in a notebook instead of just remembering them.

### File Modes

When you open a file, you tell Python *how* you want to use it:

| Mode | Meaning | File must exist? |
|------|---------|------------------|
| `'r'` | Read only | Yes |
| `'w'` | Write (overwrites!) | No — creates it |
| `'a'` | Append (adds to end) | No — creates it |
| `'x'` | Create (fails if exists) | No |
| `'r+'` | Read and write | Yes |

### The `with` Statement

Always prefer `with open(...) as file:` it automatically closes the file when the block ends, even if an error occurs. Think of it like a self-closing door. 🚪

### 4.1 Writing / Creating Files

````python
# Writing to a file creates it if it doesn't exist, overwrites if it does

# Write a single string
with open("notes.txt", "w") as file:
    file.write("Python File Handling Notes\n")
    file.write("=" * 30 + "\n")
    file.write("Created with Python!\n")

print("✅ File 'notes.txt' written successfully.")
````

````python
# Writing multiple lines at once using writelines()
# writelines() takes a list of strings each string is written as-is

shopping_list = [
    "Apples\n",
    "Bread\n",
    "Milk\n",
    "Eggs\n",
    "Cheese\n",
]

with open("shopping.txt", "w") as file:
    file.writelines(shopping_list)

print("✅ Shopping list saved to 'shopping.txt'.")
````

````python
# Appending to a file adds to the END without erasing existing content
# Great for logs!

with open("shopping.txt", "a") as file:
    file.write("Pasta\n")
    file.write("Tomatoes\n")

print("✅ Extra items appended to 'shopping.txt'.")
````

### 4.2 Reading Files

````python
# Method 1: read(): reads the ENTIRE file as one big string

with open("notes.txt", "r") as file:
    content = file.read()

print("📄 Content of notes.txt:")
print(content)
````

````python
# Method 2: readlines(): returns a LIST of lines
# Each line includes the \n newline character at the end

with open("shopping.txt", "r") as file:
    lines = file.readlines()

print(f"📋 Shopping list ({len(lines)} items):")
for i, line in enumerate(lines, 1):
    # .strip() removes the trailing \n
    print(f"  {i}. {line.strip()}")
````

````python
# Method 3: iterate directly best for large files
# Reads one line at a time memory efficient!

print("📖 Reading shopping.txt line by line:")
with open("shopping.txt", "r") as file:
    for line in file:  # Python reads line by line
        print(f"  → {line.strip()}")
````

````python
# Reading with exception handling 
# What if the file doesn't exist? Catch FileNotFoundError.

def read_file_safe(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
    except FileNotFoundError:
        print(f"❌ File '{filename}' not found.")
        return None
    except PermissionError:
        print(f"❌ No permission to read '{filename}'.")
        return None

# Test with existing and non-existing files
content = read_file_safe("notes.txt")
if content:
    print(f"✅ Read {len(content)} characters from notes.txt")

read_file_safe("does_not_exist.txt")
````

## 5. OS Module

### What is the `os` Module?

The `os` module is Python's bridge to the **operating system**. It lets you interact with files and folders (directories) in a way that works on Windows, Mac, and Linux without writing different code for each system.

Think of it as Python's **file explorer** or **terminal** but available directly inside your code.

### Most Useful `os` Functions

| Function | What it does |
|---|---|
| `os.getcwd()` | Get current working directory |
| `os.listdir(path)` | List files and folders in a directory |
| `os.mkdir(path)` | Create a new folder |
| `os.makedirs(path)` | Create folders recursively (including parents) |
| `os.remove(path)` | Delete a file |
| `os.rmdir(path)` | Delete an empty folder |
| `os.rename(old, new)` | Rename a file or folder |
| `os.path.exists(path)` | Check if file/folder exists |
| `os.path.join(...)` | Build a path safely across operating systems |

````python
import os

# 1. Where am I right now?
current_dir = os.getcwd()
print(f"📁 Current working directory: {current_dir}")
````

````python
# 2. What's in the current directory?
print("📂 Contents of current directory:")
for item in os.listdir("."):
    print(f"  - {item}")
````

````python
# 3. Create a directory
new_folder = "my_python_project"

if not os.path.exists(new_folder):      # Check before creating (avoid error)
    os.mkdir(new_folder)
    print(f"✅ Created folder: '{new_folder}'")
else:
    print(f"ℹ️  Folder '{new_folder}' already exists.")

# 4. Create nested directories in one shot with makedirs()
nested_path = os.path.join("my_python_project", "data", "raw")

os.makedirs(nested_path, exist_ok=True)  # exist_ok=True: no error if already exists
print(f"✅ Created nested path: '{nested_path}'")
````

````python
# 5. Check if something exists and what it is
paths_to_check = ["notes.txt", "shopping.txt", "ghost.txt", "my_python_project"]

print("🔍 Checking file system items:")
for path in paths_to_check:
    if os.path.exists(path):
        if os.path.isfile(path):
            size = os.path.getsize(path)  # Size in bytes
            print(f"  📄 FILE:   '{path}' ({size} bytes)")
        elif os.path.isdir(path):
            print(f"  📁 FOLDER: '{path}'")
    else:
        print(f"  ❓ NOT FOUND: '{path}'")
````

````python
# 6. Rename a file
# Let's create a file, rename it, then check

with open("old_name.txt", "w") as f:
    f.write("This file will be renamed!")

os.rename("old_name.txt", "new_name.txt")
print("✅ Renamed 'old_name.txt' → 'new_name.txt'")
print(f"   Exists now? {os.path.exists('new_name.txt')}")

# 7. Delete a file
os.remove("new_name.txt")
print("🗑️  Deleted 'new_name.txt'")
print(f"   Exists now? {os.path.exists('new_name.txt')}")
````

````python
# 8. os.path.join(): build file paths correctly
# Never concatenate paths with + or / — use os.path.join() instead!
# This makes code work on BOTH Windows (\) and Unix (/)

folder = "my_python_project"
subfolder = "data"
filename = "results.csv"

path = os.path.join(folder, subfolder, filename)
print(f"✅ Joined path: {path}")

# Useful path operations
print(f"   Directory part: {os.path.dirname(path)}")
print(f"   File name part: {os.path.basename(path)}")
print(f"   Extension:      {os.path.splitext(filename)[1]}")
````

````python
# 9. Walk a directory tree visit EVERY folder and file recursively
# os.walk() is like exploring every room and drawer in a building

# First, let's create some files to explore
os.makedirs("my_python_project/data/raw", exist_ok=True)
os.makedirs("my_python_project/data/processed", exist_ok=True)
os.makedirs("my_python_project/scripts", exist_ok=True)

for fname in ["my_python_project/data/raw/input.csv",
              "my_python_project/data/processed/output.csv",
              "my_python_project/scripts/analysis.py"]:
    with open(fname, "w") as f:
        f.write("# placeholder")

# Now walk the tree!
print("🌳 Directory tree of 'my_python_project':")
for root, dirs, files in os.walk("my_python_project"):
    # Calculate indentation based on depth
    depth = root.replace("my_python_project", "").count(os.sep)
    indent = "  " * depth
    print(f"{indent}📁 {os.path.basename(root)}/")
    for file in files:
        print(f"{indent}  📄 {file}")
````

## 6. pathlib Module

### A Modern, Cleaner Way to Work with Paths

While `os.path` works well, Python 3.4+ introduced `pathlib` a more elegant, object-oriented approach. Instead of calling functions like `os.path.join()`, you work with **Path objects** that behave like smart strings.

Think of `os.path` as a collection of old hand tools 🔧, and `pathlib` as a modern Swiss Army knife 🗡️ — same job, much more pleasant to use.

### Key `pathlib` Features

| `os.path` way | `pathlib` way | What it does |
|---|---|---|
| `os.path.join(a, b)` | `Path(a) / b` | Join paths |
| `os.path.exists(p)` | `Path(p).exists()` | Check existence |
| `os.path.basename(p)` | `Path(p).name` | Get filename |
| `os.path.dirname(p)` | `Path(p).parent` | Get parent folder |
| `os.listdir(p)` | `Path(p).iterdir()` | List contents |
| `os.makedirs(p)` | `Path(p).mkdir(parents=True)` | Create folders |

````python
from pathlib import Path

# 1. Create Path objects
current = Path(".")                   # Current directory
home = Path.home()                    # User's home directory
notes = Path("notes.txt")
project = Path("my_python_project")

print(f"📁 Current directory: {current.resolve()}")  # resolve() = absolute path
print(f"🏠 Home directory:    {home}")
````

````python
# 2. The / operator for joining paths much more readable than os.path.join!
data_dir = project / "data" / "raw"
output_file = project / "data" / "processed" / "final_report.txt"

print(f"📂 Data directory:  {data_dir}")
print(f"📄 Output file:     {output_file}")
````

````python
# 3. Path properties access parts of a path as attributes
p = Path("my_python_project/data/raw/input.csv")

print(f"Full path:      {p}")
print(f"Name:           {p.name}")       # 'input.csv'
print(f"Stem:           {p.stem}")       # 'input' (name without extension)
print(f"Suffix:         {p.suffix}")     # '.csv'
print(f"Parent:         {p.parent}")     # 'my_python_project/data/raw'
print(f"Parts:          {p.parts}")      # Tuple of all parts
print(f"Is absolute:    {p.is_absolute()}")
````

````python
# 4. Checking what exists
paths_to_check = [
    Path("notes.txt"),
    Path("my_python_project"),
    Path("ghost_file.txt"),
]

for p in paths_to_check:
    if p.exists():
        kind = "FILE 📄" if p.is_file() else "FOLDER 📁"
        print(f"  ✅ {kind}: {p}")
    else:
        print(f"  ❓ NOT FOUND: {p}")
````

````python
# 5. Reading and writing files — directly on the Path object!
# No need to open/close manually for simple cases

my_file = Path("pathlib_demo.txt")

# Write text
my_file.write_text("Hello from pathlib!\nThis is line 2.\nAnd line 3.")
print(f"✅ Written to {my_file}")

# Read text
content = my_file.read_text()
print(f"\n📄 Content:\n{content}")
````

````python
# 6. Listing directory contents
print("📂 Files in current directory:")
for item in Path(".").iterdir():
    icon = "📁" if item.is_dir() else "📄"
    print(f"  {icon} {item.name}")
````

````python
# 7. Globbing: find files matching a pattern
# glob() finds files matching a pattern (* = any characters)

print("🔍 All .txt files in current directory:")
for txt_file in Path(".").glob("*.txt"):
    print(f"  📄 {txt_file.name}")

print("\n🔍 All files in my_python_project (recursive):")
for f in Path("my_python_project").rglob("*"):  # rglob = recursive glob
    if f.is_file():
        print(f"  📄 {f}")
````

## 7. Directory Management

### Organising the File System

Managing directories (folders) programmatically is one of the most common tasks in scripting and data processing. Python gives you multiple tools to create, navigate, list, copy, move, and delete directories.

We'll combine `os`, `pathlib`, and the `shutil` module (Python's **file operations toolkit**) for a complete picture.

### `shutil` — High-Level File Operations

| Function | What it does |
|---|---|
| `shutil.copy(src, dst)` | Copy a file |
| `shutil.move(src, dst)` | Move a file or folder |
| `shutil.copytree(src, dst)` | Copy an entire folder tree |
| `shutil.rmtree(path)` | Delete a folder and everything in it |

````python
import os
import shutil
from pathlib import Path

# Create a project structure from scratch
def create_project_structure(project_name):
    """Creates a standard project folder structure."""
    base = Path(project_name)
    
    # Define the folder structure
    folders = [
        base / "src",
        base / "tests",
        base / "data" / "input",
        base / "data" / "output",
        base / "docs",
        base / "logs",
    ]
    
    # Create all folders
    for folder in folders:
        folder.mkdir(parents=True, exist_ok=True)
    
    # Create some starter files
    (base / "README.md").write_text(f"# {project_name}\n\nProject description here.")
    (base / "src" / "main.py").write_text("# Main script\nprint('Hello, World!')")
    (base / "data" / "input" / "sample.csv").write_text("name,age,city\nAlice,30,Milan")
    
    print(f"✅ Project '{project_name}' created!")
    return base


project = create_project_structure("my_data_project")
````

````python
# Visualize the directory tree
def print_tree(path, prefix=""):
    """Print a visual tree of a directory."""
    path = Path(path)
    items = sorted(path.iterdir(), key=lambda x: (x.is_file(), x.name))
    
    for i, item in enumerate(items):
        connector = "└── " if i == len(items) - 1 else "├── "
        icon = "📄" if item.is_file() else "📁"
        print(f"{prefix}{connector}{icon} {item.name}")
        
        if item.is_dir():
            extension = "    " if i == len(items) - 1 else "│   "
            print_tree(item, prefix + extension)

print("📁 my_data_project/")
print_tree("my_data_project")
````

````python
# Copying files and directories
src_file = Path("my_data_project/data/input/sample.csv")
dst_file = Path("my_data_project/data/output/sample_copy.csv")

shutil.copy(src_file, dst_file)
print(f"✅ Copied: {src_file} → {dst_file}")

# Moving a file
moved = Path("my_data_project/logs/activity.log")
moved.write_text("Log entry 1\nLog entry 2")

print(f"\n📁 Final structure:")
print("📁 my_data_project/")
print_tree("my_data_project")
````

````python
# Practical: scan a directory and report statistics
def directory_report(path):
    """Generate a report of a directory's contents."""
    path = Path(path)
    
    all_files = list(path.rglob("*"))    # Find everything recursively
    files = [f for f in all_files if f.is_file()]
    dirs  = [d for d in all_files if d.is_dir()]
    
    # Group files by extension
    by_extension = {}
    for file in files:
        ext = file.suffix or "(no extension)"
        by_extension[ext] = by_extension.get(ext, 0) + 1
    
    total_size = sum(f.stat().st_size for f in files)
    
    print(f"📊 Directory Report: {path}")
    print(f"{'='*40}")
    print(f"  📁 Subdirectories: {len(dirs)}")
    print(f"  📄 Total files:    {len(files)}")
    print(f"  💾 Total size:     {total_size} bytes")
    print(f"\n  Files by type:")
    for ext, count in sorted(by_extension.items()):
        print(f"    {ext:<20} {count} file(s)")

directory_report("my_data_project")
````

````python
# Cleanup: remove an entire directory tree
# ⚠️ WARNING: shutil.rmtree() is permanent. There's no recycle bin!

def safe_remove_tree(path):
    """Remove a directory tree with confirmation."""
    path = Path(path)
    if not path.exists():
        print(f"❓ '{path}' does not exist.")
        return
    shutil.rmtree(path)
    print(f"🗑️  Removed '{path}' and all its contents.")

# Clean up our demo projects
safe_remove_tree("my_python_project")
safe_remove_tree("my_data_project")
````

## 8. MySQL with Python

### What is MySQL?

MySQL is one of the most popular **relational databases** in the world. A database is like a super-powered Excel spreadsheet it stores data in tables (rows and columns), and you can query it with **SQL** (Structured Query Language).

**When to use MySQL instead of files?**
- You have a large amount of data (thousands to millions of records)
- Multiple users need to read/write simultaneously

````python
# ── CREATING A DATABASE AND TABLE ──────────────────────────────────

# The SQL commands we would run:
SQL_CREATE_DB = "CREATE DATABASE IF NOT EXISTS python_course;"

SQL_CREATE_TABLE = """
CREATE TABLE IF NOT EXISTS students (
    id          INT AUTO_INCREMENT PRIMARY KEY,
    name        VARCHAR(100)  NOT NULL,
    email       VARCHAR(150)  UNIQUE,
    grade       DECIMAL(4,2),
    enrolled_at TIMESTAMP     DEFAULT CURRENT_TIMESTAMP
);
"""

def setup_database(conn):
    """Create the table if it doesn't exist."""
    try:
        cursor = conn.cursor()
        cursor.execute(SQL_CREATE_TABLE)
        conn.commit()
        print("✅ Table 'students' ready.")
    except Exception as e:
        print(f"❌ Setup failed: {e}")
    finally:
        cursor.close()

# Show the SQL that would be executed
print("📋 SQL to create the table:")
print(SQL_CREATE_TABLE)
````

````python
# ── INSERT: Adding Records ─────────────────────────────────────────
# ⚠️  Always use parameterized queries (the %s placeholders)
# NEVER use string formatting — it opens you to SQL Injection attacks!

SQL_INSERT = "INSERT INTO students (name, email, grade) VALUES (%s, %s, %s);"

students_data = [
    ("Alice Rossi",    "alice@example.com",   9.2),
    ("Marco Ferrari",  "marco@example.com",   8.5),
    ("Sara Bianchi",   "sara@example.com",    7.8),
    ("Luca Verdi",     "luca@example.com",    9.7),
    ("Giulia Romano",  "giulia@example.com",  6.5),
]

def insert_students(conn, students):
    """Insert multiple students into the database."""
    try:
        cursor = conn.cursor()
        cursor.executemany(SQL_INSERT, students)   # Insert all at once
        conn.commit()                              # IMPORTANT: save changes
        print(f"✅ Inserted {cursor.rowcount} student records.")
    except mysql.connector.IntegrityError as e:
        print(f"❌ Duplicate entry: {e}")
        conn.rollback()   # Undo any partial changes
    except Exception as e:
        print(f"❌ Insert failed: {e}")
        conn.rollback()
    finally:
        cursor.close()

print("📋 SQL to insert a student:")
print(f"   {SQL_INSERT}")
print(f"\n📋 Sample data to insert:")
for student in students_data:
    print(f"   {student}")
````

````python
# ── SELECT: Querying Records ───────────────────────────────────────

SQL_SELECT_ALL = "SELECT id, name, email, grade FROM students ORDER BY grade DESC;"
SQL_SELECT_TOP = "SELECT name, grade FROM students WHERE grade >= %s ORDER BY grade DESC;"

def fetch_all_students(conn):
    """Retrieve all students from the database."""
    try:
        cursor = conn.cursor(dictionary=True)   # Returns rows as dicts!
        cursor.execute(SQL_SELECT_ALL)
        rows = cursor.fetchall()                # Get all results at once
        
        print(f"\n{'ID':<5} {'Name':<20} {'Email':<25} {'Grade':<6}")
        print("-" * 60)
        for row in rows:
            print(f"{row['id']:<5} {row['name']:<20} {row['email']:<25} {row['grade']}")
        return rows
    except Exception as e:
        print(f"❌ Query failed: {e}")
    finally:
        cursor.close()

def fetch_top_students(conn, min_grade=8.0):
    """Fetch students above a minimum grade."""
    try:
        cursor = conn.cursor(dictionary=True)
        cursor.execute(SQL_SELECT_TOP, (min_grade,))  # Safe parameterized query
        rows = cursor.fetchall()
        print(f"\n🏆 Students with grade ≥ {min_grade}:")
        for row in rows:
            print(f"   ⭐ {row['name']} — {row['grade']}")
        return rows
    except Exception as e:
        print(f"❌ Query failed: {e}")
    finally:
        cursor.close()

print("SELECT queries defined ✅")
````

````python
# ── UPDATE and DELETE ──────────────────────────────────────────────

SQL_UPDATE = "UPDATE students SET grade = %s WHERE name = %s;"
SQL_DELETE = "DELETE FROM students WHERE email = %s;"

def update_grade(conn, student_name, new_grade):
    """Update a student's grade."""
    try:
        cursor = conn.cursor()
        cursor.execute(SQL_UPDATE, (new_grade, student_name))
        conn.commit()
        if cursor.rowcount > 0:
            print(f"✅ Updated grade for '{student_name}' to {new_grade}")
        else:
            print(f"⚠️  No student named '{student_name}' found.")
    except Exception as e:
        print(f"❌ Update failed: {e}")
        conn.rollback()
    finally:
        cursor.close()

def delete_student(conn, email):
    """Remove a student by email."""
    try:
        cursor = conn.cursor()
        cursor.execute(SQL_DELETE, (email,))
        conn.commit()
        if cursor.rowcount > 0:
            print(f"🗑️  Deleted student with email '{email}'")
        else:
            print(f"⚠️  No student with email '{email}' found.")
    except Exception as e:
        print(f"❌ Delete failed: {e}")
        conn.rollback()
    finally:
        cursor.close()

print("UPDATE and DELETE functions defined ✅")
print("\n📋 Example SQL for update:")
print(f"   {SQL_UPDATE}")
print("\n📋 Example SQL for delete:")
print(f"   {SQL_DELETE}")
````

## 9. 🏆 Comprehensive Example: Student Report Card System

Let's bring **everything** together in one realistic application that uses all the topics covered:

- ✅ **Exception Handling** — safe input/output at every step
- ✅ **Built-in Exceptions** — catching `FileNotFoundError`, `ValueError`, etc.
- ✅ **User-Defined Exceptions** — custom `GradeError`, `StudentNotFoundError`
- ✅ **File Handling** — reading/writing CSV and text reports
- ✅ **OS Module** — checking and creating directories
- ✅ **pathlib** — modern path manipulation
- ✅ **Directory Management** — organising output files
- ✅ **SQLite/MySQL** — storing student records in a database

**Scenario:** A teacher needs a system to manage student grades, generate individual report cards (text files), and save everything to a structured folder.

````python
import os
import sqlite3
import shutil
from pathlib import Path
from datetime import datetime

# ── CUSTOM EXCEPTIONS ──────────────────────────────────────────────
class StudentNotFoundError(Exception):
    """Raised when a student is not in the database."""
    def __init__(self, name):
        super().__init__(f"Student '{name}' not found in the system.")

class GradeError(ValueError):
    """Raised when a grade value is invalid."""
    def __init__(self, grade):
        super().__init__(f"Grade {grade} is invalid. Must be between 0 and 10.")

class DuplicateStudentError(Exception):
    """Raised when adding a student that already exists."""
    def __init__(self, name):
        super().__init__(f"Student '{name}' already exists.")

print("✅ Custom exceptions defined.")
````

````python
# ── GRADE HELPERS ──────────────────────────────────────────────────
def validate_grade(grade):
    """Validate a grade value. Raises GradeError if invalid."""
    try:
        grade = float(grade)
    except (TypeError, ValueError):
        raise GradeError(grade)
    if not (0 <= grade <= 10):
        raise GradeError(grade)
    return round(grade, 2)

def grade_to_letter(grade):
    """Convert a numeric grade (0-10) to a letter grade."""
    if grade >= 9.0:  return "A"
    if grade >= 8.0:  return "B"
    if grade >= 7.0:  return "C"
    if grade >= 6.0:  return "D"
    return "F"

def grade_to_comment(grade):
    """Return a motivational comment based on grade."""
    if grade >= 9.0:  return "Outstanding performance! Keep it up."
    if grade >= 8.0:  return "Very good work. Aim for excellence!"
    if grade >= 7.0:  return "Good effort. A little more and you'll excel."
    if grade >= 6.0:  return "Satisfactory. Room for improvement."
    return "Needs significant improvement. Please seek extra help."

print("✅ Grade helpers defined.")
````

````python
# ── DATABASE SETUP ─────────────────────────────────────────────────
class StudentDB:
    """Manages student records in an SQLite database."""

    def __init__(self, db_path="students.db"):
        self.db_path = db_path
        self.conn = sqlite3.connect(db_path)
        self.conn.row_factory = sqlite3.Row
        self._create_tables()

    def _create_tables(self):
        self.conn.execute("""
            CREATE TABLE IF NOT EXISTS students (
                id         INTEGER PRIMARY KEY AUTOINCREMENT,
                name       TEXT    NOT NULL UNIQUE,
                class_name TEXT    NOT NULL,
                math       REAL,
                science    REAL,
                english    REAL,
                history    REAL,
                art        REAL
            )
        """)
        self.conn.commit()

    def add_student(self, name, class_name, grades):
        """Add a student with their grades. Raises DuplicateStudentError if exists."""
        validated = {subject: validate_grade(g) for subject, g in grades.items()}
        try:
            self.conn.execute(
                """INSERT INTO students (name, class_name, math, science, english, history, art)
                   VALUES (:name, :class_name, :math, :science, :english, :history, :art)""",
                {"name": name, "class_name": class_name, **validated}
            )
            self.conn.commit()
        except sqlite3.IntegrityError:
            raise DuplicateStudentError(name)

    def get_student(self, name):
        """Retrieve a student by name. Raises StudentNotFoundError if not found."""
        row = self.conn.execute(
            "SELECT * FROM students WHERE name = ?", (name,)
        ).fetchone()
        if row is None:
            raise StudentNotFoundError(name)
        return dict(row)

    def get_all_students(self):
        """Return all students as a list of dicts."""
        return [dict(row) for row in
                self.conn.execute("SELECT * FROM students ORDER BY name").fetchall()]

    def update_grade(self, name, subject, new_grade):
        """Update a single subject grade for a student."""
        self.get_student(name)          # Raise error if student doesn't exist
        valid_grade = validate_grade(new_grade)
        self.conn.execute(
            f"UPDATE students SET {subject} = ? WHERE name = ?",
            (valid_grade, name)
        )
        self.conn.commit()

    def close(self):
        self.conn.close()

print("✅ StudentDB class defined.")
````

````python
# ── REPORT CARD GENERATOR ──────────────────────────────────────────
SUBJECTS = ["math", "science", "english", "history", "art"]

def generate_report_card(student, output_dir):
    """
    Generate a text report card for a student and save it to a file.
    Uses pathlib for path management and exception handling for safety.
    """
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    # Calculate statistics
    grades = {s: student[s] for s in SUBJECTS if student[s] is not None}
    average = sum(grades.values()) / len(grades) if grades else 0
    best_subject = max(grades, key=grades.get)
    weak_subject = min(grades, key=grades.get)

    # Build the report content
    now = datetime.now().strftime("%d %B %Y, %H:%M")
    letter = grade_to_letter(average)
    comment = grade_to_comment(average)

    lines = [
        "=" * 50,
        f"    STUDENT REPORT CARD",
        f"    Academic Year 2024/2025",
        "=" * 50,
        f"  Student:  {student['name']}",
        f"  Class:    {student['class_name']}",
        f"  Date:     {now}",
        "-" * 50,
        f"  {'Subject':<15} {'Grade':>6}  {'Letter':>6}",
        "-" * 50,
    ]

    for subject, grade in grades.items():
        letter_g = grade_to_letter(grade)
        lines.append(f"  {subject.capitalize():<15} {grade:>6.1f}  {letter_g:>6}")

    lines += [
        "-" * 50,
        f"  {'AVERAGE':<15} {average:>6.2f}  {letter:>6}",
        "=" * 50,
        f"  Best subject:   {best_subject.capitalize()} ({grades[best_subject]:.1f})",
        f"  Focus on:       {weak_subject.capitalize()} ({grades[weak_subject]:.1f})",
        "-" * 50,
        f"  Teacher's comment:",
        f"  {comment}",
        "=" * 50,
        "",
    ]

    report_text = "\n".join(lines)

    # Save to file — use pathlib for the path
    safe_name = student['name'].replace(" ", "_").lower()
    file_path = output_dir / f"{safe_name}_report.txt"

    try:
        file_path.write_text(report_text, encoding="utf-8")
        return file_path, average
    except IOError as e:
        print(f"❌ Could not write report for {student['name']}: {e}")
        return None, average

print("✅ Report card generator defined.")
````

````python
# ── CLASS SUMMARY ──────────────────────────────────────────────────
def generate_class_summary(students, output_dir):
    """Generate a class-wide CSV summary file."""
    output_dir = Path(output_dir)
    summary_path = output_dir / "class_summary.csv"

    header = "Name,Class,Math,Science,English,History,Art,Average,Grade\n"
    rows = []

    for student in students:
        grades = [student[s] for s in SUBJECTS]
        avg = sum(grades) / len(grades)
        row = (
            f"{student['name']},{student['class_name']},"
            + ",".join(str(g) for g in grades)
            + f",{avg:.2f},{grade_to_letter(avg)}\n"
        )
        rows.append(row)

    try:
        with open(summary_path, "w", encoding="utf-8") as f:
            f.write(header)
            f.writelines(rows)
        return summary_path
    except IOError as e:
        print(f"❌ Could not write class summary: {e}")
        return None

print("✅ Class summary generator defined.")
````

````python
# ── MAIN PROGRAM ───────────────────────────────────────────────────
print("=" * 55)
print("   🎓 STUDENT REPORT CARD SYSTEM")
print("=" * 55 + "\n")

# Setup directories using pathlib
OUTPUT_DIR = Path("report_cards")
OUTPUT_DIR.mkdir(exist_ok=True)

# Initialise the database
db = StudentDB(":memory:")   # In-memory SQLite for demo

# Student data to load
class_data = [
    ("Alice Rossi",    "3A", {"math":9.5,"science":9.0,"english":8.5,"history":8.0,"art":9.2}),
    ("Marco Ferrari",  "3A", {"math":7.5,"science":8.0,"english":9.0,"history":7.5,"art":6.5}),
    ("Sara Bianchi",   "3B", {"math":8.5,"science":7.0,"english":7.5,"history":9.0,"art":8.0}),
    ("Luca Verdi",     "3B", {"math":5.5,"science":6.0,"english":5.0,"history":6.5,"art":7.0}),
    ("Giulia Romano",  "3A", {"math":9.8,"science":9.5,"english":9.2,"history":9.0,"art":8.5}),
    ("Davide Moretti", "3B", {"math":7.0,"science":7.5,"english":8.0,"history":7.0,"art":9.5}),
]

# Insert students — with exception handling
print("📥 Loading students into database...")
for name, cls, grades in class_data:
    try:
        db.add_student(name, cls, grades)
        print(f"  ✅ Added: {name} ({cls})")
    except DuplicateStudentError as e:
        print(f"  ⚠️  {e}")
    except GradeError as e:
        print(f"  ❌ Grade problem for {name}: {e}")

# Demonstrate updating a grade and error handling
print("\n📝 Grade updates:")
for name, subject, new_grade in [("Luca Verdi", "math", 7.0), ("Ghost Student", "math", 8.0)]:
    try:
        db.update_grade(name, subject, new_grade)
        print(f"  ✅ Updated {name}'s {subject} to {new_grade}")
    except StudentNotFoundError as e:
        print(f"  ❌ {e}")
    except GradeError as e:
        print(f"  ❌ {e}")
````

````python
# Generate individual report cards for each student
print("\n📄 Generating report cards...")
all_students = db.get_all_students()
averages = []

for student in all_students:
    path, avg = generate_report_card(student, OUTPUT_DIR)
    averages.append((student['name'], avg))
    if path:
        print(f"  📄 Saved: {path.name}  (avg: {avg:.2f}, grade: {grade_to_letter(avg)})")

# Generate class summary CSV
print("\n📊 Generating class summary...")
summary_path = generate_class_summary(all_students, OUTPUT_DIR)
if summary_path:
    print(f"  📊 Saved: {summary_path}")
````

````python
# ── FINAL OVERVIEW ─────────────────────────────────────────────────
print("\n" + "=" * 55)
print("  CLASS PERFORMANCE OVERVIEW")
print("=" * 55)

# Sort students by average grade
averages.sort(key=lambda x: x[1], reverse=True)

print(f"\n  {'Rank':<5} {'Student':<20} {'Average':>8}  {'Grade':>6}")
print("  " + "-" * 45)
for rank, (name, avg) in enumerate(averages, 1):
    medal = ["🥇", "🥈", "🥉"].pop(0) if rank <= 3 else f"#{rank} "
    print(f"  {medal:<4} {name:<20} {avg:>8.2f}  {grade_to_letter(avg):>6}")

class_avg = sum(a for _, a in averages) / len(averages)
print("  " + "-" * 45)
print(f"  {'CLASS AVERAGE':<25} {class_avg:>8.2f}  {grade_to_letter(class_avg):>6}")

# Show output files using pathlib
print(f"\n📁 Output files in '{OUTPUT_DIR}/':")
for f in sorted(OUTPUT_DIR.iterdir()):
    size = f.stat().st_size
    print(f"  📄 {f.name:<35} ({size} bytes)")

# Print one sample report card from file
sample_file = OUTPUT_DIR / "alice_rossi_report.txt"
if sample_file.exists():
    print("\n" + "=" * 55)
    print("  SAMPLE REPORT CARD (Alice Rossi)")
    print("=" * 55)
    print(sample_file.read_text())

db.close()
print("\n✅ System complete. All reports generated!")
````
