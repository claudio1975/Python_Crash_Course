## 1. What is OOP?

### The Big Idea

Imagine you're building a video game with characters, weapons, and enemies. Without OOP, you'd have hundreds of separate variables and functions flying around a mess. With OOP, you bundle **data + behaviour together** into neat, self-contained units called **objects**.

Think of it like the real world:
- A **dog** has properties (name, breed, age) and can do things (bark, fetch, sleep)
- A **car** has properties (brand, colour, speed) and can do things (accelerate, brake, honk)

OOP lets you model the world this way in code.

### The 4 Pillars of OOP:

| Pillar | Simple Explanation |
|--------|-------------------|
| **Encapsulation** | Bundling data + methods together; hiding internal details |
| **Inheritance** | A child class reuses and extends a parent class |
| **Polymorphism** | Same method name, different behaviour per class |
| **Abstraction** | Show only what's necessary; hide complex internals |

## 2. Classes and Objects

### What is a Class?

A **class** is a **blueprint** or **template**. It defines what properties and behaviours something will have.

An **object** is a **real instance** created from that blueprint.

```
CLASS (Blueprint)         OBJECTS (Real Things)
┌─────────────────┐       ┌──────────┐  ┌──────────┐
│   Dog           │  →→→  │  Fido    │  │  Rex     │
│   - name        │       │  name    │  │  name    │
│   - breed       │       │  breed   │  │  breed   │
│   - bark()      │       │  bark()  │  │  bark()  │
└─────────────────┘       └──────────┘  └──────────┘
```

Think of the class as a **cookie cutter** and objects as the **actual cookies**.

````python
# Defining a simple class
class Dog:
    # __init__ is the constructor: called automatically when creating an object
    def __init__(self, name, breed, age):
        self.name = name    # Attribute: each dog has its own name
        self.breed = breed  # Attribute: each dog has its own breed
        self.age = age      # Attribute: each dog has its own age

    def bark(self):
        print(f"{self.name} says: Woof! 🐶")

    def describe(self):
        print(f"I'm {self.name}, a {self.age}-year-old {self.breed}.")


# Creating OBJECTS from the class (each is independent)
dog1 = Dog("Fido", "Labrador", 3)
dog2 = Dog("Rex", "German Shepherd", 5)

dog1.bark()
dog2.bark()
dog1.describe()
dog2.describe()

# Each object has its OWN data
print(f"\ndog1's name: {dog1.name}")
print(f"dog2's name: {dog2.name}")
````

````python
# Class variables vs Instance variables
class Dog:
    species = "Canis lupus familiaris"  # CLASS variable: shared by ALL dogs

    def __init__(self, name, breed):
        self.name = name    # INSTANCE variable: unique to each dog
        self.breed = breed

dog1 = Dog("Fido", "Labrador")
dog2 = Dog("Rex", "Poodle")

# Class variable is the same for all
print(f"dog1 species: {dog1.species}")
print(f"dog2 species: {dog2.species}")

# Instance variables differ
print(f"dog1 name: {dog1.name}")
print(f"dog2 name: {dog2.name}")
````

**What just happened?**
- `class Dog:` defines the blueprint
- `__init__` is the **constructor** — it runs automatically when you create an object
- `dog1 = Dog("Fido", ...)` creates a real object in memory
- Each object has its **own copy** of instance variables

## 3. `self` as Default Argument

### Why Does `self` Exist?

When you have multiple objects of the same class, Python needs a way to know **which object** a method is being called on. `self` is Python's way of saying *"this specific object"*.

Imagine 100 bank accounts. When you say `account.deposit(200)`, Python uses `self` internally to say *"update the balance of THIS account, not someone else's"*.

```
account1.deposit(200)   →  Python calls  deposit(account1, 200)
account2.deposit(500)   →  Python calls  deposit(account2, 500)
                                                ↑
                                           this is 'self'
```

> 💡 `self` is **not a keyword** — it's a convention. Python passes it automatically. You must always include it as the first parameter of every method.

````python
# Demonstrating what self does
class Counter:
    def __init__(self, start=0):
        self.count = start   # 'self.count' belongs to THIS object

    def increment(self):
        self.count += 1      # Modify THIS object's count

    def reset(self):
        self.count = 0

    def get_value(self):
        return self.count


# Two INDEPENDENT counters
counter_a = Counter(0)
counter_b = Counter(100)

counter_a.increment()
counter_a.increment()
counter_a.increment()

counter_b.increment()

# self ensures each counter tracks its OWN value
print(f"Counter A: {counter_a.get_value()}")  # 3
print(f"Counter B: {counter_b.get_value()}")  # 101
````

````python
# self can also call OTHER methods of the same object
class Circle:
    PI = 3.14159

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.PI * self.radius ** 2    # Uses self.PI and self.radius

    def circumference(self):
        return 2 * self.PI * self.radius

    def describe(self):
        # self.area() calls another method on THIS same object!
        print(f"Circle with radius {self.radius}")
        print(f"  Area:          {self.area():.2f}")
        print(f"  Circumference: {self.circumference():.2f}")


c1 = Circle(5)
c2 = Circle(10)

c1.describe()
print()
c2.describe()
````

## 4. Encapsulation

### What is Encapsulation?

**Encapsulation** means bundling data (attributes) and behaviour (methods) together in a class, AND controlling who can access or modify them.

Think of a **capsule pill** 💊: the medicine is sealed inside. You can take the pill, but you can't directly touch the chemicals inside. The pill controls how you interact with its contents.

In Python, we control access with **naming conventions**:

| Naming | Convention | Access |
|--------|-----------|--------|
| `name` | Public | Accessible from anywhere |
| `_name` | Protected | Should only be used within the class and subclasses |
| `__name` | Private | Only accessible within the class (name-mangled) |

We use **getters and setters** (or Python's `@property`) to safely access private data.

````python
# Encapsulation: controlling access to data
class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner = owner              # Public: anyone can see it
        self._bank_name = "PyBank"      # Protected: internal use
        self.__balance = initial_balance # Private: hidden from outside

    # GETTER: safe way to read the private balance
    def get_balance(self):
        return self.__balance

    # SETTER: safe way to deposit money (with validation)
    def deposit(self, amount):
        if amount <= 0:
            print("❌ Deposit amount must be positive!")
        else:
            self.__balance += amount
            print(f"✅ Deposited €{amount}. New balance: €{self.__balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("❌ Withdrawal amount must be positive!")
        elif amount > self.__balance:
            print(f"❌ Insufficient funds! Balance: €{self.__balance}")
        else:
            self.__balance -= amount
            print(f"✅ Withdrew €{amount}. New balance: €{self.__balance}")


account = BankAccount("Alice", 1000)

account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # Blocked by validation
account.deposit(-100)   # Blocked by validation

print(f"\nBalance: €{account.get_balance()}")

# Trying to access private directly:
try:
    print(account.__balance)   # This will fail!
except AttributeError as e:
    print(f"\n⚠️  Cannot access private: {e}")
````

````python
# The Pythonic way: @property decorator for clean getters/setters
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius   # Private attribute

    @property
    def celsius(self):             # Getter — access like an attribute
        return self.__celsius

    @celsius.setter
    def celsius(self, value):      # Setter — with validation
        if value < -273.15:
            raise ValueError("Temperature below absolute zero is impossible!")
        self.__celsius = value

    @property
    def fahrenheit(self):          # Computed property (read-only)
        return (self.__celsius * 9 / 5) + 32

    @property
    def kelvin(self):
        return self.__celsius + 273.15


temp = Temperature(25)
print(f"Celsius:    {temp.celsius}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")
print(f"Kelvin:     {temp.kelvin} K")

# Using the setter with validation
temp.celsius = 100
print(f"\nAfter change: {temp.celsius}°C = {temp.fahrenheit}°F")

# This will raise an error
try:
    temp.celsius = -300
except ValueError as e:
    print(f"\n❌ Error: {e}")
````

**Key takeaways:**
- Encapsulation **protects data** from accidental modification
- Use `__` (double underscore) for private attributes
- Provide **public methods** to safely interact with private data
- `@property` gives you clean getter/setter syntax

## 5. Inheritance

### What is Inheritance?

**Inheritance** lets a new class (child) **reuse and extend** an existing class (parent). The child gets all of the parent's features for free, and can add its own.

Think of a **family tree**: children inherit traits from parents (eye colour, height) but also develop their own unique characteristics.

```
        Animal (Parent)
       /               \
    Dog (Child)      Cat (Child)
   /       \
Poodle   Labrador  (Grandchildren)
```

Benefits:
- ♻️ **Code reuse**: write common code once in the parent
- 🏗️ **Hierarchy**: model real-world relationships naturally
- 🔧 **Extensibility**: add features in children without touching the parent

````python
# The PARENT class (base class)
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age

    def eat(self):
        print(f"{self.name} is eating 🍽️")

    def sleep(self):
        print(f"{self.name} is sleeping 💤")

    def describe(self):
        print(f"{self.name} is a {self.age}-year-old {self.species}")


# CHILD class inherits from Animal
class Dog(Animal):      # (Animal) means 'Dog inherits from Animal'
    def __init__(self, name, breed, age):
        # super() calls the PARENT constructor — reuses its setup!
        super().__init__(name, species="Dog", age=age)
        self.breed = breed  # Dog-specific attribute

    def bark(self):         # Dog-specific method
        print(f"{self.name} says: Woof! 🐶")

    def fetch(self, item):  # Dog-specific method
        print(f"{self.name} fetches the {item}! 🎾")


class Cat(Animal):
    def __init__(self, name, indoor, age):
        super().__init__(name, species="Cat", age=age)
        self.indoor = indoor

    def meow(self):
        print(f"{self.name} says: Meow! 🐱")

    def purr(self):
        print(f"{self.name} is purring... 😸")


# Dog has BOTH Animal methods AND its own
dog = Dog("Fido", "Labrador", 3)
dog.describe()   # Inherited from Animal
dog.eat()        # Inherited from Animal
dog.bark()       # Dog's own method
dog.fetch("ball")# Dog's own method

print()

cat = Cat("Whiskers", indoor=True, age=2)
cat.describe()   # Inherited from Animal
cat.sleep()      # Inherited from Animal
cat.meow()       # Cat's own method
cat.purr()       # Cat's own method
````

````python
# Method Overriding: child replaces a parent method with its own version
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")  # Generic version


class Dog(Animal):
    def speak(self):                          # OVERRIDES parent method
        print(f"{self.name} says: Woof!")


class Cat(Animal):
    def speak(self):                          # OVERRIDES parent method
        print(f"{self.name} says: Meow!")


class Fish(Animal):
    pass  # No override — uses the parent's speak()


animals = [Dog("Rex"), Cat("Luna"), Fish("Nemo")]

for animal in animals:
    animal.speak()  # Each class responds differently
````

````python
# Multi-level inheritance: Grandchild → Child → Parent
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print(f"{self.brand} moves at {self.speed} km/h")


class Car(Vehicle):        # Inherits from Vehicle
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)
        self.doors = doors

    def honk(self):
        print(f"{self.brand}: Beep beep! 🚗")


class ElectricCar(Car):    # Inherits from Car (which inherits from Vehicle)
    def __init__(self, brand, speed, doors, battery_kw):
        super().__init__(brand, speed, doors)
        self.battery_kw = battery_kw

    def charge(self):
        print(f"{self.brand} is charging ({self.battery_kw} kW battery) ⚡")


tesla = ElectricCar("Tesla", 250, 4, 100)
tesla.move()   # From Vehicle (grandparent)
tesla.honk()   # From Car (parent)
tesla.charge() # ElectricCar's own method

print(f"\nIs tesla a Car? {isinstance(tesla, Car)}")
print(f"Is tesla a Vehicle? {isinstance(tesla, Vehicle)}")
print(f"Is tesla an ElectricCar? {isinstance(tesla, ElectricCar)}")
````

**Key takeaways:**
- Use `class Child(Parent):` to inherit
- `super().__init__(...)` calls the parent constructor
- Children can **override** parent methods
- `isinstance(obj, Class)` checks if an object is an instance of a class

## 6. Polymorphism

### What is Polymorphism?

**Polymorphism** means *"many forms"*. It's the ability to use the **same interface** (same method name) with **different types of objects**, each behaving in its own way.

Think of a **remote control** 🎮. The "play" button does the same *concept* (start playing) regardless of whether you're controlling a TV, a music player, or a DVD player — but each device responds differently.

Python supports polymorphism in two main ways:
1. **Method Overriding** — child classes provide their own version of a method
2. **Duck Typing** — if an object has the right methods, Python doesn't care what class it is

> *"If it walks like a duck and quacks like a duck, it's a duck."*

````python
# Polymorphism through method overriding
class Shape:
    def area(self):
        return 0

    def describe(self):
        print(f"I am a {self.__class__.__name__} with area {self.area():.2f}")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):    # Each shape calculates area differently
        return 3.14159 * self.radius ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


# POLYMORPHISM: same loop, same method call, different behaviour
shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]

for shape in shapes:
    shape.describe()   # Calls each shape's own area() automatically
````

````python
# Duck Typing: Python doesn't require formal inheritance for polymorphism!
class Printer:
    def print_document(self):
        print("Printing on paper... 🖨️")

class PDF:
    def print_document(self):
        print("Saving as PDF file... 📄")

class Email:
    def print_document(self):
        print("Sending via email... 📧")

# None of these inherit from each other — but all have print_document()
def send_to_output(output_device):
    output_device.print_document()  # Works on ANY object with this method!

outputs = [Printer(), PDF(), Email()]
for output in outputs:
    send_to_output(output)
````

````python
# Polymorphism with built-in operators (operator overloading)
class Vector:
    """A 2D vector that supports + and printing"""
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):      # Overloads the + operator
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):             # Overloads how the object is printed
        return f"Vector({self.x}, {self.y})"

    def __len__(self):             # Overloads len()
        return int((self.x**2 + self.y**2) ** 0.5)


v1 = Vector(2, 3)
v2 = Vector(5, 1)

v3 = v1 + v2               # Uses __add__
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v3}")
print(f"Magnitude of v3: {len(v3)}")
````

**Key takeaways:**
- Polymorphism = same interface, different behaviour
- Python's duck typing makes it natural and flexible
- **Dunder methods** (`__add__`, `__str__`, `__len__`) let you overload operators
- You can iterate over a list of different objects and call the same method on all

## 7. Abstraction

### What is Abstraction?

**Abstraction** means **hiding complex implementation details** and showing only what the user needs to know.

Think of a **car** 🚗: you press the accelerator and it goes faster. You don't need to understand combustion chemistry, fuel injection, or transmission gears. The complex internals are *abstracted away*, you just see a simple pedal.

In Python, abstraction is achieved through **Abstract Base Classes (ABCs)**:
- An **abstract class** defines a **blueprint** with required methods
- It **cannot be instantiated directly**, it's a template only
- Subclasses **must implement** all abstract methods

This enforces a **contract**: every class that inherits must provide certain methods.

````python
from abc import ABC, abstractmethod

# Abstract class: defines WHAT must exist, not HOW
class PaymentMethod(ABC):
    """Abstract blueprint for any payment method"""

    @abstractmethod
    def pay(self, amount):          # Must be implemented by subclasses
        pass

    @abstractmethod
    def validate(self):             # Must be implemented by subclasses
        pass

    def receipt(self, amount):      # Non-abstract: shared by all (no override needed)
        print(f"  🧾 Receipt: €{amount:.2f} paid via {self.__class__.__name__}")


# Concrete classes: define HOW each payment works
class CreditCard(PaymentMethod):
    def __init__(self, card_number, holder):
        self.__card_number = card_number
        self.holder = holder

    def validate(self):
        return len(str(self.__card_number)) == 16

    def pay(self, amount):
        if self.validate():
            print(f"  💳 Charging €{amount:.2f} to card ending ...{str(self.__card_number)[-4:]}")
            self.receipt(amount)
        else:
            print("  ❌ Invalid card number!")


class PayPal(PaymentMethod):
    def __init__(self, email):
        self.email = email

    def validate(self):
        return "@" in self.email

    def pay(self, amount):
        if self.validate():
            print(f"  🅿️  Paying €{amount:.2f} via PayPal ({self.email})")
            self.receipt(amount)
        else:
            print("  ❌ Invalid PayPal email!")


class CryptoPay(PaymentMethod):
    def __init__(self, wallet_address, currency="BTC"):
        self.wallet = wallet_address
        self.currency = currency

    def validate(self):
        return len(self.wallet) >= 26

    def pay(self, amount):
        if self.validate():
            print(f"  ₿ Paying €{amount:.2f} in {self.currency} from wallet {self.wallet[:8]}...")
            self.receipt(amount)
        else:
            print("  ❌ Invalid wallet address!")


# You CANNOT instantiate the abstract class:
try:
    p = PaymentMethod()
except TypeError as e:
    print(f"❌ Cannot create abstract class: {e}\n")

# You CAN use its concrete subclasses:
payments = [
    CreditCard(1234567890123456, "Alice Rossi"),
    PayPal("alice@example.com"),
    CryptoPay("1A2B3C4D5E6F7G8H9I0J1K2L3M", "ETH")
]

for payment in payments:
    payment.pay(99.99)
    print()
````

**Key takeaways:**
- Import `ABC` and `abstractmethod` from the `abc` module
- Abstract classes are **contracts**: any subclass must implement all `@abstractmethod` methods
- Abstract classes **cannot be instantiated** directly
- Non-abstract methods in the base class **are inherited as-is**
- Abstraction enforces **consistency** across related classes

## 8. Iterators

### What is an Iterator?

An **iterator** is an object that you can loop over **one item at a time**. Every time you ask for the next item, it gives you one until there are no more.

Think of a **book** 📖: you read one page at a time. You remember your place, and you always know what the next page is. That's an iterator.

Python's `for` loop works with iterators behind the scenes!

To make a custom iterator, your class needs two special methods:
- `__iter__()` — returns the iterator object itself
- `__next__()` — returns the next value (raises `StopIteration` when done)

````python
# What happens when you use a for loop?
my_list = [10, 20, 30]

# Python internally does this:
iterator = iter(my_list)      # Calls my_list.__iter__()
print(next(iterator))         # Calls iterator.__next__() → 10
print(next(iterator))         # → 20
print(next(iterator))         # → 30

try:
    print(next(iterator))     # → StopIteration (no more items)
except StopIteration:
    print("No more items!")
````

````python
# Building a custom iterator
class Countdown:
    """Iterator that counts down from n to 1"""

    def __init__(self, start):
        self.start = start
        self.current = start

    def __iter__(self):
        self.current = self.start   # Reset on each new loop
        return self                 # Must return self

    def __next__(self):
        if self.current <= 0:       # No more values
            raise StopIteration
        value = self.current
        self.current -= 1
        return value


countdown = Countdown(5)

# Works with a for loop!
print("Countdown:")
for number in countdown:
    print(f"  {number}...")
print("  🚀 Blast off!")
````

````python
# A more practical custom iterator: Fibonacci sequence
class FibonacciIterator:
    """Generate Fibonacci numbers up to a maximum value"""

    def __init__(self, max_value):
        self.max_value = max_value

    def __iter__(self):
        self.a = 0
        self.b = 1
        return self

    def __next__(self):
        if self.a > self.max_value:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b   # Fibonacci step
        return value


print("Fibonacci numbers up to 100:")
fibs = list(FibonacciIterator(100))
print(fibs)

print("\nIterating one by one:")
for fib in FibonacciIterator(50):
    print(f"  {fib}", end=" ")
print()
````

````python
# Iterators work naturally with Python built-ins
class EvenNumbers:
    """Generate even numbers up to a limit"""

    def __init__(self, limit):
        self.limit = limit

    def __iter__(self):
        self.current = 0
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 2
        return value


evens = EvenNumbers(20)

print("All evens up to 20:",     list(evens))         # Works with list()
print("Sum of evens up to 20:",  sum(EvenNumbers(20))) # Works with sum()
print("Max of evens up to 20:",  max(EvenNumbers(20))) # Works with max()
print("First 3 even numbers:",   list(zip(range(3), EvenNumbers(20))))
````

**Key takeaways:**
- An iterator implements `__iter__()` and `__next__()`
- Raise `StopIteration` when there are no more values
- Custom iterators work with `for` loops, `list()`, `sum()`, `max()`, etc.
- Python's `iter()` and `next()` functions call the dunder methods for you

## 9. 🏆 Comprehensive Example: University Management System

Now let's combine **all OOP concepts** into a realistic, complete example.

We'll build a mini **University Management System** that uses:
- ✅ Classes and Objects
- ✅ `self` as Default Argument
- ✅ Encapsulation (private attributes + @property)
- ✅ Inheritance (Person → Student / Professor)
- ✅ Polymorphism (method overriding + duck typing)
- ✅ Abstraction (abstract base class)
- ✅ Iterators (custom course roster iterator)

````python
from abc import ABC, abstractmethod

# ─────────────────────────────────────────────────────────
# ABSTRACTION: Abstract base class for all university members
# ─────────────────────────────────────────────────────────
class UniversityMember(ABC):
    university = "Python University 🎓"  # Class variable shared by all

    def __init__(self, name, member_id, age):
        self.name = name
        self._member_id = member_id   # Protected
        self.__age = age              # Private (ENCAPSULATION)

    # Encapsulation: controlled access to private age
    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value < 0 or value > 120:
            raise ValueError("Age must be between 0 and 120")
        self.__age = value

    @abstractmethod
    def get_role(self):               # Must be implemented by subclasses
        pass

    @abstractmethod
    def describe(self):               # Must be implemented by subclasses
        pass

    def __str__(self):                # POLYMORPHISM: __str__ overriding
        return f"[{self.get_role()}] {self.name} (ID: {self._member_id})"


print("✅ Abstract base class defined")
````

````python
# ─────────────────────────────────────────────────────────
# INHERITANCE: Student inherits from UniversityMember
# ─────────────────────────────────────────────────────────
class Student(UniversityMember):
    def __init__(self, name, student_id, age, degree_program):
        super().__init__(name, student_id, age)   # Call parent constructor
        self.degree_program = degree_program
        self.__grades = {}     # Private: {course_name: grade}
        self.courses = []

    # ENCAPSULATION: safe way to add grades
    def add_grade(self, course, grade):
        if not (0 <= grade <= 30):
            raise ValueError("Grade must be between 0 and 30")
        self.__grades[course] = grade

    # ENCAPSULATION: read-only access to grades
    @property
    def grades(self):
        return dict(self.__grades)  # Return a copy, not the original

    @property
    def gpa(self):
        if not self.__grades:
            return 0.0
        return sum(self.__grades.values()) / len(self.__grades)

    # POLYMORPHISM: each subclass implements get_role() differently
    def get_role(self):
        return "Student"

    # POLYMORPHISM: each subclass implements describe() differently
    def describe(self):
        print(f"  👨‍🎓 {self.name} | {self.degree_program} | Age: {self.age}")
        print(f"     GPA: {self.gpa:.1f}/30 | Courses enrolled: {len(self.courses)}")
        if self.__grades:
            print(f"     Grades: {self.__grades}")

    def enroll(self, course):
        if course.name not in self.courses:
            self.courses.append(course.name)
            course.add_student(self)
            print(f"  ✅ {self.name} enrolled in {course.name}")
        else:
            print(f"  ⚠️  {self.name} is already in {course.name}")


print("✅ Student class defined")
````

````python
# ─────────────────────────────────────────────────────────
# INHERITANCE: Professor inherits from UniversityMember
# ─────────────────────────────────────────────────────────
class Professor(UniversityMember):
    def __init__(self, name, prof_id, age, department, salary):
        super().__init__(name, prof_id, age)
        self.department = department
        self.__salary = salary   # Private: ENCAPSULATION
        self.courses_taught = []

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative")
        self.__salary = value

    def get_role(self):
        return "Professor"

    def describe(self):
        print(f"  👨‍🏫 Prof. {self.name} | {self.department} | Age: {self.age}")
        print(f"     Courses taught: {len(self.courses_taught)}")
        if self.courses_taught:
            print(f"     Teaching: {', '.join(self.courses_taught)}")

    def assign_to_course(self, course):
        self.courses_taught.append(course.name)
        course.professor = self
        print(f"  ✅ Prof. {self.name} assigned to {course.name}")


print("✅ Professor class defined")
````

````python
# ─────────────────────────────────────────────────────────
# ITERATOR: Course has a custom iterator over enrolled students
# ─────────────────────────────────────────────────────────
class Course:
    def __init__(self, name, credits, max_students=30):
        self.name = name
        self.credits = credits
        self.max_students = max_students
        self.__students = []        # Private list: ENCAPSULATION
        self.professor = None

    def add_student(self, student):
        if len(self.__students) >= self.max_students:
            print(f"  ❌ {self.name} is full!")
        else:
            self.__students.append(student)

    @property
    def student_count(self):
        return len(self.__students)

    # ITERATOR: __iter__ and __next__ on the Course itself
    def __iter__(self):
        self._index = 0
        return self

    def __next__(self):
        if self._index >= len(self.__students):
            raise StopIteration
        student = self.__students[self._index]
        self._index += 1
        return student

    def roster(self):
        prof_name = self.professor.name if self.professor else "TBD"
        print(f"\n  📋 {self.name} ({self.credits} credits) | Prof: {prof_name}")
        print(f"     Enrolled: {self.student_count}/{self.max_students} students")
        for i, student in enumerate(self, start=1):  # Uses the iterator!
            print(f"     {i:2}. {student.name:<20} GPA: {student.gpa:.1f}/30")


print("✅ Course class defined")
````

````python
# ─────────────────────────────────────────────────────────
# PUTTING IT ALL TOGETHER
# ─────────────────────────────────────────────────────────

print("=" * 60)
print(f"  Welcome to {UniversityMember.university}")
print("=" * 60)

# Create Professors
print("\n📌 Creating Faculty:")
prof_rossi  = Professor("Marco Rossi",   "P001", 48, "Computer Science", 65000)
prof_bianchi = Professor("Sara Bianchi", "P002", 52, "Mathematics",       70000)

# Create Courses
print("\n📌 Creating Courses:")
python_course = Course("Python Programming", credits=6, max_students=25)
math_course   = Course("Linear Algebra",     credits=9, max_students=20)
ai_course     = Course("AI & Machine Learning", credits=6, max_students=15)

# Assign professors
print()
prof_rossi.assign_to_course(python_course)
prof_rossi.assign_to_course(ai_course)
prof_bianchi.assign_to_course(math_course)
````

````python
# Create Students
print("\n📌 Enrolling Students:")
students_data = [
    ("Alice Ferretti",  "S001", 20, "Computer Science"),
    ("Luca Romano",     "S002", 22, "Computer Science"),
    ("Giulia Marino",   "S003", 21, "Mathematics"),
    ("Matteo Conti",    "S004", 23, "AI & Robotics"),
    ("Sofia De Luca",   "S005", 20, "Computer Science"),
]

students = [Student(name, sid, age, prog) for name, sid, age, prog in students_data]

# Enroll students in courses
print()
students[0].enroll(python_course)
students[0].enroll(math_course)
students[0].enroll(ai_course)

students[1].enroll(python_course)
students[1].enroll(ai_course)

students[2].enroll(math_course)
students[2].enroll(python_course)

students[3].enroll(ai_course)
students[3].enroll(python_course)

students[4].enroll(math_course)
students[4].enroll(ai_course)
````

````python
# ── POLYMORPHISM: same describe() call on different objects ──
print("\n" + "=" * 60)
print("  UNIVERSITY MEMBERS DIRECTORY")
print("=" * 60)

print("\n── Faculty ──")
all_members = [prof_rossi, prof_bianchi]  # All are UniversityMember
for member in all_members:
    print(member)          # Uses __str__ → calls get_role() polymorphically
    member.describe()      # Each class has its own describe()
    print()

print("── Students ──")
for student in students:
    print(student)         # Same interface, Student's version
    student.describe()
    print()
````

````python
# ── ITERATOR: loop over each course's roster ──
print("=" * 60)
print("  COURSE ROSTERS")
print("=" * 60)

for course in [python_course, math_course, ai_course]:
    course.roster()   # Internally uses the Course iterator!
````
